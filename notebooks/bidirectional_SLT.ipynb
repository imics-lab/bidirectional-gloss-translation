```python
#aslg with flan t5
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class GlossToTextTranslationPipeline:
    def __init__(self, model_name="google/flan-t5-base"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])

        # Ensure string type
        df[gloss_col] = df[gloss_col].astype(str)
        df[text_col] = df[text_col].astype(str)

        # Limit size for memory efficiency
        if len(df) > 2000:
            df = df.sample(2000, random_state=42).reset_index(drop=True)
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(gloss, text):
            # T5 expects input and target separately
            input_text = f"translate gloss to english: {gloss}"
            target_text = text
            return input_text, target_text

        # Create input and target columns
        df[['input_text', 'target_text']] = df.apply(
            lambda r: pd.Series(create_input_target(r[gloss_col], r[text_col])), axis=1
        )

        train_df, val_df = train_test_split(df[['input_text', 'target_text']], test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # Load T5 model
        self.model = AutoModelForSeq2SeqLM.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # LoRA config for T5
        lora_cfg = LoraConfig(
            r=16,
            lora_alpha=32,
            target_modules=["q", "v", "k", "o", "wi", "wo"],
            lora_dropout=0.1,
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./gloss_to_text_t5"):
        # Use DataCollatorForSeq2Seq for T5
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=3,
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=2,
            eval_strategy='steps',
            eval_steps=100,
            save_steps=100,
            logging_steps=25,
            learning_rate=5e-4,
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_translation(self, gloss, max_new_tokens=50):
        input_text = f"translate gloss to english: {gloss}"

        inputs = self.tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # T5 generation
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=False,
                early_stopping=True,
                no_repeat_ngram_size=2
            )

        # Decode the generated tokens
        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        return generated_text.strip()

    def generate_translations(self, glosses, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_translation(g, max_new_tokens) for g in tqdm(glosses, desc="Generating translations")]

    def calculate_metrics(self, preds, refs):
        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = GlossToTextTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation
    samples = df[gloss_col].sample(3, random_state=42).tolist()
    outputs = pipeline.generate_translations(samples)
    for gloss, out in zip(samples, outputs):
        print(f"Gloss: {gloss} -> Text: {out}")

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_glosses = val_df[gloss_col].tolist()
    val_refs    = val_df[text_col].tolist()
    val_preds   = pipeline.generate_translations(val_glosses)
    metrics = pipeline.calculate_metrics(val_preds, val_refs)
    print("Validation Metrics:")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")
```

```python
#aslg with flan t5 - improved with preprocessing
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer
import re
import json
import spacy
from collections import defaultdict

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class GlossPreprocessor:
    def __init__(self):
        # Load spaCy model for NLP processing
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except OSError:
            print("Warning: spaCy English model not found. Some preprocessing features will be limited.")
            self.nlp = None

        # Gender mapping for common names (can be expanded)
        self.gender_mapping = {
            'mr': 'mr', 'mrs': 'mrs', 'ms': 'ms', 'madam': 'madam',
            'president': 'president', 'commissioner': 'commissioner',
            'baroness': 'baroness', 'chairman': 'chairman', 'chairwoman': 'chairwoman'
        }

        # Common EU institutional terms
        self.institutional_terms = {
            'parliament', 'commission', 'council', 'european', 'union', 'eu',
            'committee', 'directive', 'regulation', 'treaty', 'member', 'state'
        }

        # Number patterns
        self.number_patterns = {
            'percentage': re.compile(r'(\d+)\s*%'),
            'year': re.compile(r'\b(19|20)\d{2}\b'),
            'decimal': re.compile(r'\d+\.\d+'),
            'currency': re.compile(r'EUR\s*\d+(?:\.\d+)?(?:\s*(?:billion|million))?', re.IGNORECASE)
        }

    def normalize_gloss_tokens(self, gloss_text):
        """Normalize ASL gloss notation"""
        if not isinstance(gloss_text, str):
            return ""

        # Convert to uppercase for consistent processing
        text = gloss_text.upper()

        # Handle possessive markers
        text = re.sub(r'X-POSS', "'S", text)

        # Handle pronoun references
        pronoun_map = {
            'X-I': 'I',
            'X-WE': 'WE',
            'X-YOU': 'YOU',
            'X-HE': 'HE',
            'X-SHE': 'SHE',
            'X-IT': 'IT',
            'X-Y': 'THEY',
            'X-MY': 'MY',
            'SE': 'THESE'
        }

        for gloss_pronoun, english_pronoun in pronoun_map.items():
            text = text.replace(gloss_pronoun, english_pronoun)

        # Handle descriptive markers
        text = re.sub(r'DESC-([A-Z]+)', r'\1', text)

        # Handle negation
        text = text.replace('DESC-NOT', 'NOT')

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def add_articles_heuristically(self, text):
        """Add articles based on linguistic patterns"""
        if not self.nlp:
            return text

        doc = self.nlp(text)
        tokens = []

        for i, token in enumerate(doc):
            # Add article before singular countable nouns
            if (token.pos_ == 'NOUN' and
                token.tag_ in ['NN'] and  # Singular noun
                i > 0 and
                doc[i-1].pos_ not in ['DET', 'PRON'] and
                token.text.lower() not in self.institutional_terms):

                # Use "an" before vowel sounds, "a" otherwise
                if token.text[0].lower() in 'aeiou':
                    tokens.append('an')
                else:
                    tokens.append('a')

            # Add "the" before superlatives and specific references
            elif (token.pos_ == 'ADJ' and token.tag_ in ['JJS'] and  # Superlative
                  i == 0 or doc[i-1].pos_ != 'DET'):
                tokens.append('the')

            tokens.append(token.text)

        return ' '.join(tokens)

    def normalize_numbers_and_dates(self, text):
        """Normalize number and date formats"""
        # Normalize percentages
        text = re.sub(r'(\d+)\s*%', r'\1 %', text)

        # Normalize years
        text = re.sub(r'\b(\d{4})\b', lambda m: m.group(1), text)

        # Normalize currency
        text = re.sub(r'EUR\s*(\d+(?:\.\d+)?)', r'eur \1', text, flags=re.IGNORECASE)

        return text

    def fix_common_errors(self, text):
        """Fix common systematic errors"""
        # Fix spacing issues
        text = re.sub(r'(\w)([,.])', r'\1 \2', text)  # Add space before punctuation
        text = re.sub(r'\s+([,.!?])', r'\1', text)    # Remove space before punctuation

        # Fix common word boundaries
        text = re.sub(r'inthe\b', 'in the', text, flags=re.IGNORECASE)
        text = re.sub(r'onthe\b', 'on the', text, flags=re.IGNORECASE)
        text = re.sub(r'tothe\b', 'to the', text, flags=re.IGNORECASE)

        # Normalize contractions
        text = re.sub(r"can't", "cannot", text, flags=re.IGNORECASE)
        text = re.sub(r"won't", "will not", text, flags=re.IGNORECASE)

        return text

    def process_gloss(self, gloss_text):
        """Main preprocessing pipeline for gloss text"""
        text = self.normalize_gloss_tokens(gloss_text)
        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)
        return text.lower()  # Convert to lowercase for consistency

    def process_target_text(self, target_text):
        """Preprocessing for target text"""
        if not isinstance(target_text, str):
            return ""

        text = target_text.strip()
        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Normalize punctuation spacing
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s+', ' ', text)

        return text.strip()

class PostProcessor:
    def __init__(self):
        # Common gender corrections based on names
        self.name_gender_corrections = {
            'angelilli': 'mrs', 'lamassoure': 'mr', 'bowles': 'mrs',
            'pagano': 'mrs', 'scheele': 'ms', 'ashton': 'baroness'
        }

        # Common institutional corrections
        self.institutional_corrections = {
            'commissions': 'commission',
            'parliaments': 'parliament',
        }

    def fix_gender_references(self, text):
        """Fix gender references based on name context"""
        words = text.split()
        corrected_words = []

        for i, word in enumerate(words):
            if word.lower() in ['mr', 'mrs', 'ms'] and i + 1 < len(words):
                name = words[i + 1].lower()
                if name in self.name_gender_corrections:
                    corrected_words.append(self.name_gender_corrections[name])
                else:
                    corrected_words.append(word)
            else:
                corrected_words.append(word)

        return ' '.join(corrected_words)

    def fix_articles(self, text):
        """Post-process to fix article usage"""
        # Add "the" before specific institutional terms
        institutional_patterns = [
            (r'\b(commission|parliament|council|union|budget|report|debate|directive)\b',
             r'the \1'),
            (r'\b(european union)\b', r'the \1'),
        ]

        for pattern, replacement in institutional_patterns:
            # Only add "the" if not already present
            text = re.sub(f'(?<!the )(?<!a )(?<!an ){pattern}', replacement, text, flags=re.IGNORECASE)

        return text

    def normalize_punctuation(self, text):
        """Normalize punctuation and spacing"""
        # Ensure proper spacing around punctuation
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s*!\s*', '!', text)
        text = re.sub(r'\s*\?\s*', '?', text)

        # Remove trailing punctuation spaces
        text = text.rstrip(' .,!?')

        return text

    def post_process(self, text):
        """Main post-processing pipeline"""
        text = self.fix_gender_references(text)
        text = self.fix_articles(text)
        text = self.normalize_punctuation(text)

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

class GlossToTextTranslationPipeline:
    def __init__(self, model_name="google/flan-t5-base"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

        # Initialize preprocessors
        self.gloss_preprocessor = GlossPreprocessor()
        self.post_processor = PostProcessor()

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])

        # Apply preprocessing to both gloss and text columns
        print("Applying gloss preprocessing...")
        df[gloss_col] = df[gloss_col].apply(self.gloss_preprocessor.process_gloss)

        print("Applying target text preprocessing...")
        df[text_col] = df[text_col].apply(self.gloss_preprocessor.process_target_text)

        # Filter out very short or very long sequences
        df = df[df[gloss_col].str.len() > 10]  # Remove very short glosses
        df = df[df[text_col].str.len() > 10]   # Remove very short texts
        df = df[df[gloss_col].str.len() < 500] # Remove very long glosses
        df = df[df[text_col].str.len() < 500]  # Remove very long texts

        # Limit size for memory efficiency
        if len(df) > 2000:
            df = df.sample(2000, random_state=42).reset_index(drop=True)

        print(f"Dataset after preprocessing: {len(df)} samples")
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(gloss, text):
            # Enhanced prompt for T5
            input_text = f"translate american sign language gloss to english text: {gloss}"
            target_text = text
            return input_text, target_text

        # Create input and target columns
        df[['input_text', 'target_text']] = df.apply(
            lambda r: pd.Series(create_input_target(r[gloss_col], r[text_col])), axis=1
        )

        # Data augmentation: create variations with different prompts
        augmented_data = []
        for _, row in df.iterrows():
            # Original
            augmented_data.append({
                'input_text': row['input_text'],
                'target_text': row['target_text']
            })

            # Alternative prompt
            augmented_data.append({
                'input_text': f"convert gloss to text: {row[gloss_col]}",
                'target_text': row['target_text']
            })

        augmented_df = pd.DataFrame(augmented_data)
        train_df, val_df = train_test_split(augmented_df, test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # Load T5 model
        self.model = AutoModelForSeq2SeqLM.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # Enhanced LoRA config for T5
        lora_cfg = LoraConfig(
            r=32,  # Increased rank for better performance
            lora_alpha=64,
            target_modules=["q", "v", "k", "o", "wi", "wo"],
            lora_dropout=0.05,  # Reduced dropout
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./gloss_to_text_t5_improved"):
        # Use DataCollatorForSeq2Seq for T5
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=5,  # Increased epochs
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=4,  # Increased for better gradients
            eval_strategy='steps',
            eval_steps=50,
            save_steps=50,
            logging_steps=25,
            learning_rate=3e-4,  # Slightly reduced learning rate
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True,
            warmup_steps=100,  # Added warmup
            weight_decay=0.01  # Added weight decay
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_translation(self, gloss, max_new_tokens=50):
        # Preprocess the gloss
        processed_gloss = self.gloss_preprocessor.process_gloss(gloss)
        input_text = f"translate american sign language gloss to english text: {processed_gloss}"

        inputs = self.tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # T5 generation with improved parameters
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=True,  # Enable sampling
                temperature=0.7,  # Add temperature for diversity
                top_p=0.9,       # Add nucleus sampling
                early_stopping=True,
                no_repeat_ngram_size=3,  # Increased to avoid repetition
                length_penalty=1.0
            )

        # Decode the generated tokens
        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Apply post-processing
        processed_text = self.post_processor.post_process(generated_text.strip())

        return processed_text

    def generate_translations(self, glosses, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_translation(g, max_new_tokens) for g in tqdm(glosses, desc="Generating translations")]

    def calculate_metrics(self, preds, refs):
        # Apply post-processing to predictions
        processed_preds = [self.post_processor.post_process(pred) for pred in preds]

        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=processed_preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=processed_preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=processed_preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(processed_preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = GlossToTextTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # Show some preprocessing examples
    print("\nPreprocessing examples:")
    for i in range(min(3, len(df))):
        print(f"Original gloss: {raw_ds['train'][i][gloss_col]}")
        print(f"Processed gloss: {df.iloc[i][gloss_col]}")
        print(f"Target text: {df.iloc[i][text_col]}")
        print("-" * 50)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    print(f"Training samples: {len(tok_ds['train'])}")
    print(f"Validation samples: {len(tok_ds['validation'])}")

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation
    samples = df[gloss_col].sample(5, random_state=42).tolist()
    outputs = pipeline.generate_translations(samples)
    print("\nSample translations:")
    for gloss, out in zip(samples, outputs):
        print(f"Gloss: {gloss}")
        print(f"Generated: {out}")
        print("-" * 50)

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_glosses = val_df[gloss_col].tolist()
    val_refs    = val_df[text_col].tolist()
    val_preds   = pipeline.generate_translations(val_glosses)
    metrics = pipeline.calculate_metrics(val_preds, val_refs)

    print("Validation Metrics:")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")

    # Show some prediction vs reference examples
    print("\nSample predictions vs references:")
    for i in range(min(5, len(val_preds))):
        print(f"Reference: {val_refs[i]}")
        print(f"Predicted: {val_preds[i]}")
        print("-" * 50)
```

```python
#aslg with t5 - improved with preprocessing
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer
import re
import json
import spacy
from collections import defaultdict

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class GlossPreprocessor:
    def __init__(self):
        # Load spaCy model for NLP processing
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except OSError:
            print("Warning: spaCy English model not found. Some preprocessing features will be limited.")
            self.nlp = None

        # Gender mapping for common names (can be expanded)
        self.gender_mapping = {
            'mr': 'mr', 'mrs': 'mrs', 'ms': 'ms', 'madam': 'madam',
            'president': 'president', 'commissioner': 'commissioner',
            'baroness': 'baroness', 'chairman': 'chairman', 'chairwoman': 'chairwoman'
        }

        # Common EU institutional terms
        self.institutional_terms = {
            'parliament', 'commission', 'council', 'european', 'union', 'eu',
            'committee', 'directive', 'regulation', 'treaty', 'member', 'state'
        }

        # Number patterns
        self.number_patterns = {
            'percentage': re.compile(r'(\d+)\s*%'),
            'year': re.compile(r'\b(19|20)\d{2}\b'),
            'decimal': re.compile(r'\d+\.\d+'),
            'currency': re.compile(r'EUR\s*\d+(?:\.\d+)?(?:\s*(?:billion|million))?', re.IGNORECASE)
        }

    def normalize_gloss_tokens(self, gloss_text):
        """Normalize ASL gloss notation"""
        if not isinstance(gloss_text, str):
            return ""

        # Convert to uppercase for consistent processing
        text = gloss_text.upper()

        # Handle possessive markers
        text = re.sub(r'X-POSS', "'S", text)

        # Handle pronoun references
        pronoun_map = {
            'X-I': 'I',
            'X-WE': 'WE',
            'X-YOU': 'YOU',
            'X-HE': 'HE',
            'X-SHE': 'SHE',
            'X-IT': 'IT',
            'X-Y': 'THEY',
            'X-MY': 'MY',
            'SE': 'THESE'
        }

        for gloss_pronoun, english_pronoun in pronoun_map.items():
            text = text.replace(gloss_pronoun, english_pronoun)

        # Handle descriptive markers
        text = re.sub(r'DESC-([A-Z]+)', r'\1', text)

        # Handle negation
        text = text.replace('DESC-NOT', 'NOT')

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def add_articles_heuristically(self, text):
        """Add articles based on linguistic patterns"""
        if not self.nlp:
            return text

        doc = self.nlp(text)
        tokens = []

        for i, token in enumerate(doc):
            # Add article before singular countable nouns
            if (token.pos_ == 'NOUN' and
                token.tag_ in ['NN'] and  # Singular noun
                i > 0 and
                doc[i-1].pos_ not in ['DET', 'PRON'] and
                token.text.lower() not in self.institutional_terms):

                # Use "an" before vowel sounds, "a" otherwise
                if token.text[0].lower() in 'aeiou':
                    tokens.append('an')
                else:
                    tokens.append('a')

            # Add "the" before superlatives and specific references
            elif (token.pos_ == 'ADJ' and token.tag_ in ['JJS'] and  # Superlative
                  i == 0 or doc[i-1].pos_ != 'DET'):
                tokens.append('the')

            tokens.append(token.text)

        return ' '.join(tokens)

    def normalize_numbers_and_dates(self, text):
        """Normalize number and date formats"""
        # Normalize percentages
        text = re.sub(r'(\d+)\s*%', r'\1 %', text)

        # Normalize years
        text = re.sub(r'\b(\d{4})\b', lambda m: m.group(1), text)

        # Normalize currency
        text = re.sub(r'EUR\s*(\d+(?:\.\d+)?)', r'eur \1', text, flags=re.IGNORECASE)

        return text

    def fix_common_errors(self, text):
        """Fix common systematic errors"""
        # Fix spacing issues
        text = re.sub(r'(\w)([,.])', r'\1 \2', text)  # Add space before punctuation
        text = re.sub(r'\s+([,.!?])', r'\1', text)    # Remove space before punctuation

        # Fix common word boundaries
        text = re.sub(r'inthe\b', 'in the', text, flags=re.IGNORECASE)
        text = re.sub(r'onthe\b', 'on the', text, flags=re.IGNORECASE)
        text = re.sub(r'tothe\b', 'to the', text, flags=re.IGNORECASE)

        # Normalize contractions
        text = re.sub(r"can't", "cannot", text, flags=re.IGNORECASE)
        text = re.sub(r"won't", "will not", text, flags=re.IGNORECASE)

        return text

    def process_gloss(self, gloss_text):
        """Main preprocessing pipeline for gloss text"""
        text = self.normalize_gloss_tokens(gloss_text)
        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)
        return text.lower()  # Convert to lowercase for consistency

    def process_target_text(self, target_text):
        """Preprocessing for target text"""
        if not isinstance(target_text, str):
            return ""

        text = target_text.strip()
        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Normalize punctuation spacing
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s+', ' ', text)

        return text.strip()

class PostProcessor:
    def __init__(self):
        # Common gender corrections based on names
        self.name_gender_corrections = {
            'angelilli': 'mrs', 'lamassoure': 'mr', 'bowles': 'mrs',
            'pagano': 'mrs', 'scheele': 'ms', 'ashton': 'baroness'
        }

        # Common institutional corrections
        self.institutional_corrections = {
            'commissions': 'commission',
            'parliaments': 'parliament',
        }

    def fix_gender_references(self, text):
        """Fix gender references based on name context"""
        words = text.split()
        corrected_words = []

        for i, word in enumerate(words):
            if word.lower() in ['mr', 'mrs', 'ms'] and i + 1 < len(words):
                name = words[i + 1].lower()
                if name in self.name_gender_corrections:
                    corrected_words.append(self.name_gender_corrections[name])
                else:
                    corrected_words.append(word)
            else:
                corrected_words.append(word)

        return ' '.join(corrected_words)

    def fix_articles(self, text):
        """Post-process to fix article usage"""
        # Add "the" before specific institutional terms
        institutional_patterns = [
            (r'\b(commission|parliament|council|union|budget|report|debate|directive)\b',
             r'the \1'),
            (r'\b(european union)\b', r'the \1'),
        ]

        for pattern, replacement in institutional_patterns:
            # Only add "the" if not already present
            text = re.sub(f'(?<!the )(?<!a )(?<!an ){pattern}', replacement, text, flags=re.IGNORECASE)

        return text

    def normalize_punctuation(self, text):
        """Normalize punctuation and spacing"""
        # Ensure proper spacing around punctuation
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s*!\s*', '!', text)
        text = re.sub(r'\s*\?\s*', '?', text)

        # Remove trailing punctuation spaces
        text = text.rstrip(' .,!?')

        return text

    def post_process(self, text):
        """Main post-processing pipeline"""
        text = self.fix_gender_references(text)
        text = self.fix_articles(text)
        text = self.normalize_punctuation(text)

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

class GlossToTextTranslationPipeline:
    def __init__(self, model_name="t5-base"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

        # Initialize preprocessors
        self.gloss_preprocessor = GlossPreprocessor()
        self.post_processor = PostProcessor()

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])

        # Apply preprocessing to both gloss and text columns
        print("Applying gloss preprocessing...")
        df[gloss_col] = df[gloss_col].apply(self.gloss_preprocessor.process_gloss)

        print("Applying target text preprocessing...")
        df[text_col] = df[text_col].apply(self.gloss_preprocessor.process_target_text)

        # Filter out very short or very long sequences
        df = df[df[gloss_col].str.len() > 10]  # Remove very short glosses
        df = df[df[text_col].str.len() > 10]   # Remove very short texts
        df = df[df[gloss_col].str.len() < 500] # Remove very long glosses
        df = df[df[text_col].str.len() < 500]  # Remove very long texts

        # Limit size for memory efficiency
        if len(df) > 2000:
            df = df.sample(2000, random_state=42).reset_index(drop=True)

        print(f"Dataset after preprocessing: {len(df)} samples")
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(gloss, text):
            # T5 expects task prefix for sequence-to-sequence tasks
            input_text = f"translate american sign language gloss to english text: {gloss}"
            target_text = text
            return input_text, target_text

        # Create input and target columns
        df[['input_text', 'target_text']] = df.apply(
            lambda r: pd.Series(create_input_target(r[gloss_col], r[text_col])), axis=1
        )

        # Data augmentation: create variations with different prompts
        augmented_data = []
        for _, row in df.iterrows():
            # Original
            augmented_data.append({
                'input_text': row['input_text'],
                'target_text': row['target_text']
            })

            # Alternative prompt
            augmented_data.append({
                'input_text': f"convert gloss to text: {row[gloss_col]}",
                'target_text': row['target_text']
            })

        augmented_df = pd.DataFrame(augmented_data)
        train_df, val_df = train_test_split(augmented_df, test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        # Load T5 tokenizer and model
        self.tokenizer = T5Tokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # Load T5 model
        self.model = T5ForConditionalGeneration.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # Enhanced LoRA config for T5
        lora_cfg = LoraConfig(
            r=32,  # Increased rank for better performance
            lora_alpha=64,
            target_modules=["q", "v", "k", "o", "wi", "wo"],
            lora_dropout=0.05,  # Reduced dropout
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./gloss_to_text_t5_improved"):
        # Use DataCollatorForSeq2Seq for T5
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=5,  # Increased epochs
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=4,  # Increased for better gradients
            eval_strategy='steps',
            eval_steps=50,
            save_steps=50,
            logging_steps=25,
            learning_rate=3e-4,  # Slightly reduced learning rate
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True,
            warmup_steps=100,  # Added warmup
            weight_decay=0.01  # Added weight decay
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_translation(self, gloss, max_new_tokens=50):
        # Preprocess the gloss
        processed_gloss = self.gloss_preprocessor.process_gloss(gloss)
        input_text = f"translate american sign language gloss to english text: {processed_gloss}"

        inputs = self.tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # T5 generation with improved parameters
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=True,  # Enable sampling
                temperature=0.7,  # Add temperature for diversity
                top_p=0.9,       # Add nucleus sampling
                early_stopping=True,
                no_repeat_ngram_size=3,  # Increased to avoid repetition
                length_penalty=1.0
            )

        # Decode the generated tokens
        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Apply post-processing
        processed_text = self.post_processor.post_process(generated_text.strip())

        return processed_text

    def generate_translations(self, glosses, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_translation(g, max_new_tokens) for g in tqdm(glosses, desc="Generating translations")]

    def calculate_metrics(self, preds, refs):
        # Apply post-processing to predictions
        processed_preds = [self.post_processor.post_process(pred) for pred in preds]

        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=processed_preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=processed_preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=processed_preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(processed_preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = GlossToTextTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # Show some preprocessing examples
    print("\nPreprocessing examples:")
    for i in range(min(3, len(df))):
        print(f"Original gloss: {raw_ds['train'][i][gloss_col]}")
        print(f"Processed gloss: {df.iloc[i][gloss_col]}")
        print(f"Target text: {df.iloc[i][text_col]}")
        print("-" * 50)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    print(f"Training samples: {len(tok_ds['train'])}")
    print(f"Validation samples: {len(tok_ds['validation'])}")

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation
    samples = df[gloss_col].sample(5, random_state=42).tolist()
    outputs = pipeline.generate_translations(samples)
    print("\nSample translations:")
    for gloss, out in zip(samples, outputs):
        print(f"Gloss: {gloss}")
        print(f"Generated: {out}")
        print("-" * 50)

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_glosses = val_df[gloss_col].tolist()
    val_refs    = val_df[text_col].tolist()
    val_preds   = pipeline.generate_translations(val_glosses)
    metrics = pipeline.calculate_metrics(val_preds, val_refs)

    print("Validation Metrics:")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")

    # Show some prediction vs reference examples
    print("\nSample predictions vs references:")
    for i in range(min(5, len(val_preds))):
        print(f"Reference: {val_refs[i]}")
        print(f"Predicted: {val_preds[i]}")
        print("-" * 50)
```

```python
#aslg with t5 - improved with preprocessing
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer
import re
import json
import spacy
from collections import defaultdict

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class GlossPreprocessor:
    def __init__(self):
        # Load spaCy model for NLP processing
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except OSError:
            print("Warning: spaCy English model not found. Some preprocessing features will be limited.")
            self.nlp = None

        # Gender mapping for common names (can be expanded)
        self.gender_mapping = {
            'mr': 'mr', 'mrs': 'mrs', 'ms': 'ms', 'madam': 'madam',
            'president': 'president', 'commissioner': 'commissioner',
            'baroness': 'baroness', 'chairman': 'chairman', 'chairwoman': 'chairwoman'
        }

        # Common EU institutional terms
        self.institutional_terms = {
            'parliament', 'commission', 'council', 'european', 'union', 'eu',
            'committee', 'directive', 'regulation', 'treaty', 'member', 'state'
        }

        # Number patterns
        self.number_patterns = {
            'percentage': re.compile(r'(\d+)\s*%'),
            'year': re.compile(r'\b(19|20)\d{2}\b'),
            'decimal': re.compile(r'\d+\.\d+'),
            'currency': re.compile(r'EUR\s*\d+(?:\.\d+)?(?:\s*(?:billion|million))?', re.IGNORECASE)
        }

    def normalize_gloss_tokens(self, gloss_text):
        """Normalize ASL gloss notation"""
        if not isinstance(gloss_text, str):
            return ""

        # Remove BOM and clean whitespace
        text = gloss_text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to uppercase for consistent processing
        text = text.upper()

        # Handle possessive markers
        text = re.sub(r'X-POSS', "'S", text)

        # Handle pronoun references
        pronoun_map = {
            'X-I': 'I',
            'X-WE': 'WE',
            'X-YOU': 'YOU',
            'X-HE': 'HE',
            'X-SHE': 'SHE',
            'X-IT': 'IT',
            'X-THEY': 'THEY',
            'X-MY': 'MY',
            'SE': 'THESE'
        }

        for gloss_pronoun, english_pronoun in pronoun_map.items():
            text = text.replace(gloss_pronoun, english_pronoun)

        # Handle descriptive markers
        text = re.sub(r'DESC-([A-Z]+)', r'\1', text)

        # Handle negation
        text = text.replace('DESC-NOT', 'NOT')

        # Handle common ASL grammatical markers
        text = re.sub(r'\bSEE\s+MINUTE\b', '', text)  # Remove "SEE MINUTE"
        text = re.sub(r'\bMEMBERSHIP\s+PARLIAMENT\b', '', text)  # Remove structural markers
        text = re.sub(r'\bAPPROVAL\s+MINUTE\b', '', text)  # Remove approval markers
        text = re.sub(r'\bDESC-PREVIOUS\s+SIT\b', '', text)  # Remove descriptive markers

        # Clean up extra spaces and return
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def add_articles_heuristically(self, text):
        """Add articles based on linguistic patterns"""
        if not self.nlp:
            return text

        doc = self.nlp(text)
        tokens = []

        for i, token in enumerate(doc):
            # Add article before singular countable nouns
            if (token.pos_ == 'NOUN' and
                token.tag_ in ['NN'] and  # Singular noun
                i > 0 and
                doc[i-1].pos_ not in ['DET', 'PRON'] and
                token.text.lower() not in self.institutional_terms):

                # Use "an" before vowel sounds, "a" otherwise
                if token.text[0].lower() in 'aeiou':
                    tokens.append('an')
                else:
                    tokens.append('a')

            # Add "the" before superlatives and specific references
            elif (token.pos_ == 'ADJ' and token.tag_ in ['JJS'] and  # Superlative
                  i == 0 or doc[i-1].pos_ != 'DET'):
                tokens.append('the')

            tokens.append(token.text)

        return ' '.join(tokens)

    def normalize_numbers_and_dates(self, text):
        """Normalize number and date formats"""
        # Normalize percentages
        text = re.sub(r'(\d+)\s*%', r'\1 %', text)

        # Normalize years
        text = re.sub(r'\b(\d{4})\b', lambda m: m.group(1), text)

        # Normalize currency
        text = re.sub(r'EUR\s*(\d+(?:\.\d+)?)', r'eur \1', text, flags=re.IGNORECASE)

        return text

    def fix_common_errors(self, text):
        """Fix common systematic errors"""
        # Fix spacing issues
        text = re.sub(r'(\w)([,.])', r'\1 \2', text)  # Add space before punctuation
        text = re.sub(r'\s+([,.!?])', r'\1', text)    # Remove space before punctuation

        # Fix common word boundaries
        text = re.sub(r'inthe\b', 'in the', text, flags=re.IGNORECASE)
        text = re.sub(r'onthe\b', 'on the', text, flags=re.IGNORECASE)
        text = re.sub(r'tothe\b', 'to the', text, flags=re.IGNORECASE)

        # Normalize contractions
        text = re.sub(r"can't", "cannot", text, flags=re.IGNORECASE)
        text = re.sub(r"won't", "will not", text, flags=re.IGNORECASE)

        return text

    def process_gloss(self, gloss_text):
        """Main preprocessing pipeline for gloss text"""
        if not isinstance(gloss_text, str) or not gloss_text.strip():
            return ""

        text = self.normalize_gloss_tokens(gloss_text)

        # Skip further processing if text is empty after normalization
        if not text.strip():
            return ""

        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Convert to lowercase for consistency, but preserve some structure
        text = text.lower()

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def process_target_text(self, target_text):
        """Preprocessing for target text"""
        if not isinstance(target_text, str):
            return ""

        text = target_text.strip()
        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Normalize punctuation spacing
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s+', ' ', text)

        return text.strip()

class PostProcessor:
    def __init__(self):
        # Common gender corrections based on names
        self.name_gender_corrections = {
            'angelilli': 'mrs', 'lamassoure': 'mr', 'bowles': 'mrs',
            'pagano': 'mrs', 'scheele': 'ms', 'ashton': 'baroness'
        }

        # Common institutional corrections
        self.institutional_corrections = {
            'commissions': 'commission',
            'parliaments': 'parliament',
        }

    def fix_gender_references(self, text):
        """Fix gender references based on name context"""
        words = text.split()
        corrected_words = []

        for i, word in enumerate(words):
            if word.lower() in ['mr', 'mrs', 'ms'] and i + 1 < len(words):
                name = words[i + 1].lower()
                if name in self.name_gender_corrections:
                    corrected_words.append(self.name_gender_corrections[name])
                else:
                    corrected_words.append(word)
            else:
                corrected_words.append(word)

        return ' '.join(corrected_words)

    def fix_articles(self, text):
        """Post-process to fix article usage"""
        # Add "the" before specific institutional terms
        institutional_patterns = [
            (r'\b(commission|parliament|council|union|budget|report|debate|directive)\b',
             r'the \1'),
            (r'\b(european union)\b', r'the \1'),
        ]

        for pattern, replacement in institutional_patterns:
            # Only add "the" if not already present
            text = re.sub(f'(?<!the )(?<!a )(?<!an ){pattern}', replacement, text, flags=re.IGNORECASE)

        return text

    def normalize_punctuation(self, text):
        """Normalize punctuation and spacing"""
        # Ensure proper spacing around punctuation
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s*!\s*', '!', text)
        text = re.sub(r'\s*\?\s*', '?', text)

        # Remove trailing punctuation spaces
        text = text.rstrip(' .,!?')

        return text

    def post_process(self, text):
        """Main post-processing pipeline"""
        text = self.fix_gender_references(text)
        text = self.fix_articles(text)
        text = self.normalize_punctuation(text)

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

class GlossToTextTranslationPipeline:
    def __init__(self, model_name="t5-base"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

        # Initialize preprocessors
        self.gloss_preprocessor = GlossPreprocessor()
        self.post_processor = PostProcessor()

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        print(f"Found columns: gloss='{gloss_col}', text='{text_col}'")
        print(f"Original dataset size: {len(df)}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])
        print(f"After dropping nulls: {len(df)}")

        # Create copies to preserve original data
        df = df.copy()
        original_gloss = df[gloss_col].copy()
        original_text = df[text_col].copy()

        # Apply preprocessing to both gloss and text columns
        print("Applying gloss preprocessing...")
        df[gloss_col] = df[gloss_col].apply(self.gloss_preprocessor.process_gloss)

        print("Applying target text preprocessing...")
        df[text_col] = df[text_col].apply(self.gloss_preprocessor.process_target_text)

        # Filter out very short or very long sequences
        valid_mask = (
            (df[gloss_col].str.len() > 10) &  # Remove very short glosses
            (df[text_col].str.len() > 10) &   # Remove very short texts
            (df[gloss_col].str.len() < 500) & # Remove very long glosses
            (df[text_col].str.len() < 500)    # Remove very long texts
        )

        df = df[valid_mask].reset_index(drop=True)
        original_gloss = original_gloss[valid_mask].reset_index(drop=True)
        original_text = original_text[valid_mask].reset_index(drop=True)

        print(f"After length filtering: {len(df)}")

        # Limit size for memory efficiency
        if len(df) > 2000:
            sample_indices = df.sample(2000, random_state=42).index
            df = df.loc[sample_indices].reset_index(drop=True)
            original_gloss = original_gloss.loc[sample_indices].reset_index(drop=True)
            original_text = original_text.loc[sample_indices].reset_index(drop=True)

        # Add original columns for debugging
        df['original_gloss'] = original_gloss
        df['original_text'] = original_text

        print(f"Final dataset size: {len(df)} samples")
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(gloss, text):
            # T5 expects task prefix for sequence-to-sequence tasks
            input_text = f"translate american sign language gloss to english text: {gloss}"
            target_text = text
            return input_text, target_text

        # Create input and target columns
        df[['input_text', 'target_text']] = df.apply(
            lambda r: pd.Series(create_input_target(r[gloss_col], r[text_col])), axis=1
        )

        # Data augmentation: create variations with different prompts
        augmented_data = []
        for _, row in df.iterrows():
            # Original
            augmented_data.append({
                'input_text': row['input_text'],
                'target_text': row['target_text']
            })

            # Alternative prompt
            augmented_data.append({
                'input_text': f"convert gloss to text: {row[gloss_col]}",
                'target_text': row['target_text']
            })

        augmented_df = pd.DataFrame(augmented_data)
        train_df, val_df = train_test_split(augmented_df, test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        # Load T5 tokenizer and model
        self.tokenizer = T5Tokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # Load T5 model
        self.model = T5ForConditionalGeneration.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # Enhanced LoRA config for T5
        lora_cfg = LoraConfig(
            r=32,  # Increased rank for better performance
            lora_alpha=64,
            target_modules=["q", "v", "k", "o", "wi", "wo"],
            lora_dropout=0.05,  # Reduced dropout
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./gloss_to_text_t5_improved"):
        # Use DataCollatorForSeq2Seq for T5
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=5,  # Increased epochs
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=4,  # Increased for better gradients
            eval_strategy='steps',
            eval_steps=50,
            save_steps=50,
            logging_steps=25,
            learning_rate=3e-4,  # Slightly reduced learning rate
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True,
            warmup_steps=100,  # Added warmup
            weight_decay=0.01  # Added weight decay
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_translation(self, gloss, max_new_tokens=50):
        # Preprocess the gloss
        processed_gloss = self.gloss_preprocessor.process_gloss(gloss)
        input_text = f"translate american sign language gloss to english text: {processed_gloss}"

        inputs = self.tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # T5 generation with improved parameters
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=True,  # Enable sampling
                temperature=0.7,  # Add temperature for diversity
                top_p=0.9,       # Add nucleus sampling
                early_stopping=True,
                no_repeat_ngram_size=3,  # Increased to avoid repetition
                length_penalty=1.0
            )

        # Decode the generated tokens
        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Apply post-processing
        processed_text = self.post_processor.post_process(generated_text.strip())

        return processed_text

    def generate_translations(self, glosses, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_translation(g, max_new_tokens) for g in tqdm(glosses, desc="Generating translations")]

    def calculate_metrics(self, preds, refs):
        # Apply post-processing to predictions
        processed_preds = [self.post_processor.post_process(pred) for pred in preds]

        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=processed_preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=processed_preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=processed_preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(processed_preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = GlossToTextTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # Show some preprocessing examples with original data
    print("\nPreprocessing examples:")
    for i in range(min(3, len(df))):
        print(f"Original gloss: {df.iloc[i]['original_gloss']}")
        print(f"Processed gloss: {df.iloc[i][gloss_col]}")
        print(f"Original text: {df.iloc[i]['original_text']}")
        print(f"Processed text: {df.iloc[i][text_col]}")
        print("-" * 50)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    print(f"Training samples: {len(tok_ds['train'])}")
    print(f"Validation samples: {len(tok_ds['validation'])}")

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation
    samples = df[gloss_col].sample(5, random_state=42).tolist()
    outputs = pipeline.generate_translations(samples)
    print("\nSample translations:")
    for gloss, out in zip(samples, outputs):
        print(f"Gloss: {gloss}")
        print(f"Generated: {out}")
        print("-" * 50)

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_glosses = val_df[gloss_col].tolist()
    val_refs    = val_df[text_col].tolist()
    val_preds   = pipeline.generate_translations(val_glosses)
    metrics = pipeline.calculate_metrics(val_preds, val_refs)

    print("Validation Metrics:")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")

    # Show some prediction vs reference examples with original context
    print("\nSample predictions vs references:")
    for i in range(min(5, len(val_preds))):
        # Find corresponding original data
        original_idx = val_df.index[i]
        print(f"Original gloss: {df.loc[original_idx, 'original_gloss']}")
        print(f"Processed gloss: {val_glosses[i]}")
        print(f"Reference: {val_refs[i]}")
        print(f"Predicted: {val_preds[i]}")
        print("-" * 50)
```

```python
#aslg with mbart-50-small - improved with preprocessing
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    MBartTokenizer,
    MBartForConditionalGeneration,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer
import re
import json
import spacy
from collections import defaultdict

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class GlossPreprocessor:
    def __init__(self):
        # Load spaCy model for NLP processing
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except OSError:
            print("Warning: spaCy English model not found. Some preprocessing features will be limited.")
            self.nlp = None

        # Gender mapping for common names (can be expanded)
        self.gender_mapping = {
            'mr': 'mr', 'mrs': 'mrs', 'ms': 'ms', 'madam': 'madam',
            'president': 'president', 'commissioner': 'commissioner',
            'baroness': 'baroness', 'chairman': 'chairman', 'chairwoman': 'chairwoman'
        }

        # Common EU institutional terms
        self.institutional_terms = {
            'parliament', 'commission', 'council', 'european', 'union', 'eu',
            'committee', 'directive', 'regulation', 'treaty', 'member', 'state'
        }

        # Number patterns
        self.number_patterns = {
            'percentage': re.compile(r'(\d+)\s*%'),
            'year': re.compile(r'\b(19|20)\d{2}\b'),
            'decimal': re.compile(r'\d+\.\d+'),
            'currency': re.compile(r'EUR\s*\d+(?:\.\d+)?(?:\s*(?:billion|million))?', re.IGNORECASE)
        }

    def normalize_gloss_tokens(self, gloss_text):
        """Normalize ASL gloss notation"""
        if not isinstance(gloss_text, str):
            return ""

        # Remove BOM and clean whitespace
        text = gloss_text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to uppercase for consistent processing
        text = text.upper()

        # Handle possessive markers
        text = re.sub(r'X-POSS', "'S", text)

        # Handle pronoun references
        pronoun_map = {
            'X-I': 'I',
            'X-WE': 'WE',
            'X-YOU': 'YOU',
            'X-HE': 'HE',
            'X-SHE': 'SHE',
            'X-IT': 'IT',
            'X-THEY': 'THEY',
            'X-MY': 'MY',
            'SE': 'THESE'
        }

        for gloss_pronoun, english_pronoun in pronoun_map.items():
            text = text.replace(gloss_pronoun, english_pronoun)

        # Handle descriptive markers
        text = re.sub(r'DESC-([A-Z]+)', r'\1', text)

        # Handle negation
        text = text.replace('DESC-NOT', 'NOT')

        # Handle common ASL grammatical markers
        text = re.sub(r'\bSEE\s+MINUTE\b', '', text)  # Remove "SEE MINUTE"
        text = re.sub(r'\bMEMBERSHIP\s+PARLIAMENT\b', '', text)  # Remove structural markers
        text = re.sub(r'\bAPPROVAL\s+MINUTE\b', '', text)  # Remove approval markers
        text = re.sub(r'\bDESC-PREVIOUS\s+SIT\b', '', text)  # Remove descriptive markers

        # Clean up extra spaces and return
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def add_articles_heuristically(self, text):
        """Add articles based on linguistic patterns"""
        if not self.nlp:
            return text

        doc = self.nlp(text)
        tokens = []

        for i, token in enumerate(doc):
            # Add article before singular countable nouns
            if (token.pos_ == 'NOUN' and
                token.tag_ in ['NN'] and  # Singular noun
                i > 0 and
                doc[i-1].pos_ not in ['DET', 'PRON'] and
                token.text.lower() not in self.institutional_terms):

                # Use "an" before vowel sounds, "a" otherwise
                if token.text[0].lower() in 'aeiou':
                    tokens.append('an')
                else:
                    tokens.append('a')

            # Add "the" before superlatives and specific references
            elif (token.pos_ == 'ADJ' and token.tag_ in ['JJS'] and  # Superlative
                  i == 0 or doc[i-1].pos_ != 'DET'):
                tokens.append('the')

            tokens.append(token.text)

        return ' '.join(tokens)

    def normalize_numbers_and_dates(self, text):
        """Normalize number and date formats"""
        # Normalize percentages
        text = re.sub(r'(\d+)\s*%', r'\1 %', text)

        # Normalize years
        text = re.sub(r'\b(\d{4})\b', lambda m: m.group(1), text)

        # Normalize currency
        text = re.sub(r'EUR\s*(\d+(?:\.\d+)?)', r'eur \1', text, flags=re.IGNORECASE)

        return text

    def fix_common_errors(self, text):
        """Fix common systematic errors"""
        # Fix spacing issues
        text = re.sub(r'(\w)([,.])', r'\1 \2', text)  # Add space before punctuation
        text = re.sub(r'\s+([,.!?])', r'\1', text)    # Remove space before punctuation

        # Fix common word boundaries
        text = re.sub(r'inthe\b', 'in the', text, flags=re.IGNORECASE)
        text = re.sub(r'onthe\b', 'on the', text, flags=re.IGNORECASE)
        text = re.sub(r'tothe\b', 'to the', text, flags=re.IGNORECASE)

        # Normalize contractions
        text = re.sub(r"can't", "cannot", text, flags=re.IGNORECASE)
        text = re.sub(r"won't", "will not", text, flags=re.IGNORECASE)

        return text

    def process_gloss(self, gloss_text):
        """Main preprocessing pipeline for gloss text"""
        if not isinstance(gloss_text, str) or not gloss_text.strip():
            return ""

        text = self.normalize_gloss_tokens(gloss_text)

        # Skip further processing if text is empty after normalization
        if not text.strip():
            return ""

        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Convert to lowercase for consistency, but preserve some structure
        text = text.lower()

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def process_target_text(self, target_text):
        """Preprocessing for target text"""
        if not isinstance(target_text, str):
            return ""

        text = target_text.strip()
        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Normalize punctuation spacing
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s+', ' ', text)

        return text.strip()

class PostProcessor:
    def __init__(self):
        # Common gender corrections based on names
        self.name_gender_corrections = {
            'angelilli': 'mrs', 'lamassoure': 'mr', 'bowles': 'mrs',
            'pagano': 'mrs', 'scheele': 'ms', 'ashton': 'baroness'
        }

        # Common institutional corrections
        self.institutional_corrections = {
            'commissions': 'commission',
            'parliaments': 'parliament',
        }

    def fix_gender_references(self, text):
        """Fix gender references based on name context"""
        words = text.split()
        corrected_words = []

        for i, word in enumerate(words):
            if word.lower() in ['mr', 'mrs', 'ms'] and i + 1 < len(words):
                name = words[i + 1].lower()
                if name in self.name_gender_corrections:
                    corrected_words.append(self.name_gender_corrections[name])
                else:
                    corrected_words.append(word)
            else:
                corrected_words.append(word)

        return ' '.join(corrected_words)

    def fix_articles(self, text):
        """Post-process to fix article usage"""
        # Add "the" before specific institutional terms
        institutional_patterns = [
            (r'\b(commission|parliament|council|union|budget|report|debate|directive)\b',
             r'the \1'),
            (r'\b(european union)\b', r'the \1'),
        ]

        for pattern, replacement in institutional_patterns:
            # Only add "the" if not already present
            text = re.sub(f'(?<!the )(?<!a )(?<!an ){pattern}', replacement, text, flags=re.IGNORECASE)

        return text

    def normalize_punctuation(self, text):
        """Normalize punctuation and spacing"""
        # Ensure proper spacing around punctuation
        text = re.sub(r'\s*,\s*', ', ', text)
        text = re.sub(r'\s*\.\s*', '. ', text)
        text = re.sub(r'\s*!\s*', '!', text)
        text = re.sub(r'\s*\?\s*', '?', text)

        # Remove trailing punctuation spaces
        text = text.rstrip(' .,!?')

        return text

    def post_process(self, text):
        """Main post-processing pipeline"""
        text = self.fix_gender_references(text)
        text = self.fix_articles(text)
        text = self.normalize_punctuation(text)

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

class GlossToTextTranslationPipeline:
    def __init__(self, model_name="facebook/mbart-large-50-many-to-many-mmt"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

        # Initialize preprocessors
        self.gloss_preprocessor = GlossPreprocessor()
        self.post_processor = PostProcessor()

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        print(f"Found columns: gloss='{gloss_col}', text='{text_col}'")
        print(f"Original dataset size: {len(df)}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])
        print(f"After dropping nulls: {len(df)}")

        # Create copies to preserve original data
        df = df.copy()
        original_gloss = df[gloss_col].copy()
        original_text = df[text_col].copy()

        # Apply preprocessing to both gloss and text columns
        print("Applying gloss preprocessing...")
        df[gloss_col] = df[gloss_col].apply(self.gloss_preprocessor.process_gloss)

        print("Applying target text preprocessing...")
        df[text_col] = df[text_col].apply(self.gloss_preprocessor.process_target_text)

        # Filter out very short or very long sequences
        valid_mask = (
            (df[gloss_col].str.len() > 10) &  # Remove very short glosses
            (df[text_col].str.len() > 10) &   # Remove very short texts
            (df[gloss_col].str.len() < 500) & # Remove very long glosses
            (df[text_col].str.len() < 500)    # Remove very long texts
        )

        df = df[valid_mask].reset_index(drop=True)
        original_gloss = original_gloss[valid_mask].reset_index(drop=True)
        original_text = original_text[valid_mask].reset_index(drop=True)

        print(f"After length filtering: {len(df)}")

        # Limit size for memory efficiency
        if len(df) > 2000:
            sample_indices = df.sample(2000, random_state=42).index
            df = df.loc[sample_indices].reset_index(drop=True)
            original_gloss = original_gloss.loc[sample_indices].reset_index(drop=True)
            original_text = original_text.loc[sample_indices].reset_index(drop=True)

        # Add original columns for debugging
        df['original_gloss'] = original_gloss
        df['original_text'] = original_text

        print(f"Final dataset size: {len(df)} samples")
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(gloss, text):
            # mBART expects source text without task prefix
            input_text = gloss
            target_text = text
            return input_text, target_text

        # Create input and target columns
        df[['input_text', 'target_text']] = df.apply(
            lambda r: pd.Series(create_input_target(r[gloss_col], r[text_col])), axis=1
        )

        # Data augmentation can be simpler for mBART
        augmented_data = []
        for _, row in df.iterrows():
            augmented_data.append({
                'input_text': row['input_text'],
                'target_text': row['target_text']
            })

        augmented_df = pd.DataFrame(augmented_data)
        train_df, val_df = train_test_split(augmented_df, test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        # Load mBART tokenizer and model
        self.tokenizer = MBartTokenizer.from_pretrained(self.model_name)

        # mBART uses specific language codes
        self.tokenizer.src_lang = "en_XX"  # Source language (we'll treat gloss as English variant)
        self.tokenizer.tgt_lang = "en_XX"  # Target language

        # Load mBART model
        self.model = MBartForConditionalGeneration.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # Enhanced LoRA config for mBART
        lora_cfg = LoraConfig(
            r=32,  # Increased rank for better performance
            lora_alpha=64,
            target_modules=["q_proj", "v_proj", "k_proj", "out_proj", "fc1", "fc2"],
            lora_dropout=0.05,  # Reduced dropout
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs with source language
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length,
                add_special_tokens=True
            )

            # Tokenize targets with target language
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length,
                    add_special_tokens=True
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./gloss_to_text_mbart_improved"):
        # Use DataCollatorForSeq2Seq for mBART
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=5,  # Increased epochs
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=4,  # Increased for better gradients
            eval_strategy='steps',
            eval_steps=50,
            save_steps=50,
            logging_steps=25,
            learning_rate=3e-4,  # Slightly reduced learning rate
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True,
            warmup_steps=100,  # Added warmup
            weight_decay=0.01  # Added weight decay
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_translation(self, gloss, max_new_tokens=50):
        # Preprocess the gloss
        processed_gloss = self.gloss_preprocessor.process_gloss(gloss)

        # Set source language and tokenize
        self.tokenizer.src_lang = "en_XX"
        inputs = self.tokenizer(
            processed_gloss,
            return_tensors="pt",
            truncation=True,
            max_length=256,
            add_special_tokens=True
        ).to(self.peft_model.device)

        # mBART generation with improved parameters
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=True,  # Enable sampling
                temperature=0.7,  # Add temperature for diversity
                top_p=0.9,       # Add nucleus sampling
                early_stopping=True,
                no_repeat_ngram_size=3,  # Increased to avoid repetition
                length_penalty=1.0,
                forced_bos_token_id=self.tokenizer.lang_code_to_id["en_XX"]  # Force target language
            )

        # Decode the generated tokens
        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Apply post-processing
        processed_text = self.post_processor.post_process(generated_text.strip())

        return processed_text

    def generate_translations(self, glosses, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_translation(g, max_new_tokens) for g in tqdm(glosses, desc="Generating translations")]

    def calculate_metrics(self, preds, refs):
        # Apply post-processing to predictions
        processed_preds = [self.post_processor.post_process(pred) for pred in preds]

        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=processed_preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=processed_preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=processed_preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(processed_preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = GlossToTextTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # Show some preprocessing examples with original data
    print("\nPreprocessing examples:")
    for i in range(min(3, len(df))):
        print(f"Original gloss: {df.iloc[i]['original_gloss']}")
        print(f"Processed gloss: {df.iloc[i][gloss_col]}")
        print(f"Original text: {df.iloc[i]['original_text']}")
        print(f"Processed text: {df.iloc[i][text_col]}")
        print("-" * 50)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    print(f"Training samples: {len(tok_ds['train'])}")
    print(f"Validation samples: {len(tok_ds['validation'])}")

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation
    samples = df[gloss_col].sample(5, random_state=42).tolist()
    outputs = pipeline.generate_translations(samples)
    print("\nSample translations:")
    for gloss, out in zip(samples, outputs):
        print(f"Gloss: {gloss}")
        print(f"Generated: {out}")
        print("-" * 50)

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_glosses = val_df[gloss_col].tolist()
    val_refs    = val_df[text_col].tolist()
    val_preds   = pipeline.generate_translations(val_glosses)
    metrics = pipeline.calculate_metrics(val_preds, val_refs)

    print("Validation Metrics:")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")

    # Show some prediction vs reference examples with original context
    print("\nSample predictions vs references:")
    for i in range(min(5, len(val_preds))):
        # Find corresponding original data
        original_idx = val_df.index[i]
        print(f"Original gloss: {df.loc[original_idx, 'original_gloss']}")
        print(f"Processed gloss: {val_glosses[i]}")
        print(f"Reference: {val_refs[i]}")
        print(f"Predicted: {val_preds[i]}")
        print("-" * 50)
```

```python
import pandas as pd
from datasets import Dataset
import torch
from transformers import (
    T5ForConditionalGeneration,
    T5Tokenizer,
    Seq2SeqTrainingArguments,
    Seq2SeqTrainer,
    EarlyStoppingCallback,
    DataCollatorForSeq2Seq
)
from sklearn.metrics import accuracy_score
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction
from sklearn.model_selection import train_test_split
import nltk
from nltk.tokenize import word_tokenize
import random
import numpy as np
import logging
import jiwer
from rouge_score import rouge_scorer

# ─── Setup ─────────────────────────────────────────────────────────────────────
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    handlers=[logging.StreamHandler(), logging.FileHandler("training.log")]
)
logger = logging.getLogger(__name__)

nltk.download("punkt")
nltk.download("wordnet")

# ─── Helpers ───────────────────────────────────────────────────────────────────
def clean_text(text):
    return " ".join(text.split()).lower().replace("_", " ")

def calculate_bleu_scores(reference, hypothesis):
    r = reference.lower().split()
    h = hypothesis.lower().split()
    smoothie = SmoothingFunction().method1
    weights_list = [
        (1,0,0,0),
        (0.5,0.5,0,0),
        (1/3,1/3,1/3,0),
        (0.25,0.25,0.25,0.25),
    ]
    return {
        f"bleu_{i}": corpus_bleu([[r]], [h], weights=w, smoothing_function=smoothie)
        for i, w in enumerate(weights_list, start=1)
    }

def calculate_wer(reference, hypothesis):
    return jiwer.wer(reference.lower(), hypothesis.lower())

def calculate_rouge_scores(reference, hypothesis):
    scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
    sc = scorer.score(reference, hypothesis)
    return {
        'rouge1_fmeasure': sc['rouge1'].fmeasure,
        'rouge2_fmeasure': sc['rouge2'].fmeasure,
        'rougeL_fmeasure': sc['rougeL'].fmeasure
    }

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    logger.info(f"Seed set to {seed}")

# ─── Main ──────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    set_seed()

    # Load and merge data
    anno_path, trans_path = 'signum_sents_anno_eng.txt', 'signum_sents_trans_eng.txt'
    gloss_data = [dict(zip(['id','gloss'], line.strip().split('\t'))) for line in open(anno_path)]
    trans_data = [dict(zip(['id','translation'], line.strip().split('\t'))) for line in open(trans_path)]
    df = pd.merge(pd.DataFrame(gloss_data), pd.DataFrame(trans_data), on='id')
    df['gloss'] = df['gloss'].apply(clean_text)
    df['translation'] = df['translation'].apply(clean_text)
    logger.info(f"{len(df)} examples loaded")

    # Augment
    def add_noise(text, p=0.1):
        words = text.split()
        if random.random() < p and len(words) > 3:
            words.pop(random.randrange(len(words)))
        return ' '.join(words)

    rows = []
    for _, row in df.iterrows():
        rows.append(row.to_dict())
        if random.random() < 0.3:
            rows.append({
                'id': row['id'] + '_aug',
                'gloss': add_noise(row['gloss']),
                'translation': row['translation']
            })
    df = pd.DataFrame(rows)
    logger.info(f"After augmentation: {len(df)} examples")

    # Split
    train_df, tmp_df = train_test_split(df, test_size=0.2, random_state=42)
    eval_df, test_df = train_test_split(tmp_df, test_size=0.5, random_state=42)
    logger.info(f"Splits — train: {len(train_df)}, eval: {len(eval_df)}, test: {len(test_df)}")

    train_ds = Dataset.from_pandas(train_df)
    eval_ds  = Dataset.from_pandas(eval_df)
    test_ds  = Dataset.from_pandas(test_df)

    # Model & Tokenizer (T5-Base, slow tokenizer)
    model_name = "t5-base"
    tokenizer  = T5Tokenizer.from_pretrained(model_name, use_fast=False)
    model      = T5ForConditionalGeneration.from_pretrained(model_name)
    device     = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Preprocessing
    def preprocess_fn(examples):
        inputs = [f"translate Sign Language gloss to English: {g}" for g in examples["gloss"]]
        m = tokenizer(inputs, max_length=128, truncation=True, padding="max_length")
        lbl = tokenizer(examples["translation"], max_length=128, truncation=True, padding="max_length")
        m["labels"] = [
            [(l if l != tokenizer.pad_token_id else -100) for l in seq]
            for seq in lbl["input_ids"]
        ]
        return m

    train_ds = train_ds.map(preprocess_fn, batched=True, remove_columns=["id","gloss","translation"])
    eval_ds  = eval_ds.map(preprocess_fn,  batched=True, remove_columns=["id","gloss","translation"])
    test_ds  = test_ds.map(preprocess_fn,  batched=True, remove_columns=["id","gloss","translation"])
    train_ds.set_format("torch")
    eval_ds.set_format("torch")
    test_ds.set_format("torch")

    data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, padding=True, max_length=128)

    # Training arguments
    bs = 4 if torch.cuda.is_available() else 2
    training_args = Seq2SeqTrainingArguments(
        output_dir="./results_t5_base",
        eval_strategy="steps",
        eval_steps=50,
        save_steps=100,
        save_total_limit=2,
        num_train_epochs=3,
        learning_rate=3e-4,
        per_device_train_batch_size=bs,
        per_device_eval_batch_size=bs,
        gradient_accumulation_steps=2,
        warmup_ratio=0.1,
        weight_decay=0.01,
        logging_dir="./logs_t5_base",
        logging_steps=10,
        report_to="none",
        predict_with_generate=True,
        generation_max_length=128,
        load_best_model_at_end=True,
        metric_for_best_model="bleu_4",
        greater_is_better=True,
        fp16=torch.cuda.is_available(),
    )

    # Filter out-of-range IDs before decoding
    vocab_size = len(tokenizer.get_vocab())

    def compute_metrics(eval_pred):
        preds, labels = eval_pred

        # if logits, take argmax
        if isinstance(preds, np.ndarray) and preds.ndim == 3:
            preds = np.argmax(preds, axis=-1)

        # replace -100 in labels
        labels = np.where(labels != -100, labels, tokenizer.pad_token_id)

        # to pure Python int lists, filter invalid IDs
        pred_seqs, label_seqs = [], []
        for seq in preds.tolist():
            pred_seqs.append([int(x) for x in seq if 0 <= int(x) < vocab_size])
        for seq in labels.tolist():
            label_seqs.append([int(x) for x in seq if 0 <= int(x) < vocab_size])

        # decode one by one
        dec_preds  = [tokenizer.decode(s, skip_special_tokens=True).strip() for s in pred_seqs]
        dec_labels = [tokenizer.decode(s, skip_special_tokens=True).strip() for s in label_seqs]

        # compute BLEU, WER, ROUGE
        bleu1,bleu2,bleu3,bleu4 = [],[],[],[]
        wers = []
        r1,r2,rl = [],[],[]
        for ref, hyp in zip(dec_labels, dec_preds):
            bs = calculate_bleu_scores(ref, hyp)
            bleu1.append(bs["bleu_1"])
            bleu2.append(bs["bleu_2"])
            bleu3.append(bs["bleu_3"])
            bleu4.append(bs["bleu_4"])
            wers.append(calculate_wer(ref, hyp))
            rs = calculate_rouge_scores(ref, hyp)
            r1.append(rs["rouge1_fmeasure"])
            r2.append(rs["rouge2_fmeasure"])
            rl.append(rs["rougeL_fmeasure"])

        return {
            "bleu_1":          float(np.mean(bleu1)),
            "bleu_2":          float(np.mean(bleu2)),
            "bleu_3":          float(np.mean(bleu3)),
            "bleu_4":          float(np.mean(bleu4)),
            "wer":             float(np.mean(wers)),
            "rouge1_fmeasure": float(np.mean(r1)),
            "rouge2_fmeasure": float(np.mean(r2)),
            "rougeL_fmeasure": float(np.mean(rl))
        }

    # Trainer
    trainer = Seq2SeqTrainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=eval_ds,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]
    )

    # Train & Evaluate
    logger.info("Starting training…")
    trainer.train()
    logger.info("Training completed")
    metrics = trainer.evaluate(test_ds)
    logger.info(f"Test metrics: {metrics}")

    # Save
    model.save_pretrained("./fine_tuned_t5_base")
    tokenizer.save_pretrained("./fine_tuned_t5_base")
    logger.info("Model & tokenizer saved to ./fine_tuned_t5_base")
```

```python
#t5 with intense
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class GlossToTextTranslationPipeline:
    def __init__(self, model_name="t5-base"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

    def load_dataset(self, dataset_name="merterm/intensified-phoenix-14-t"):
        """Load intensified-phoenix-14-t dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Use specific column names
        gloss_col = 'English Gloss'
        text_col = 'English Text'

        # Check if columns exist
        if gloss_col not in df.columns or text_col not in df.columns:
            raise ValueError(f"Dataset must have columns '{gloss_col}' and '{text_col}'. Available columns: {df.columns.tolist()}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])

        # Ensure string type
        df[gloss_col] = df[gloss_col].astype(str)
        df[text_col] = df[text_col].astype(str)

        # Limit size for memory efficiency
        if len(df) > 2000:
            df = df.sample(2000, random_state=42).reset_index(drop=True)
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(gloss, text):
            # T5 expects input and target separately
            input_text = f"translate english gloss to english text: {gloss}"
            target_text = text
            return input_text, target_text

        # Create input and target columns
        df[['input_text', 'target_text']] = df.apply(
            lambda r: pd.Series(create_input_target(r[gloss_col], r[text_col])), axis=1
        )

        train_df, val_df = train_test_split(df[['input_text', 'target_text']], test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # Load T5 model
        self.model = AutoModelForSeq2SeqLM.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # LoRA config for T5
        lora_cfg = LoraConfig(
            r=16,
            lora_alpha=32,
            target_modules=["q", "v", "k", "o", "wi", "wo"],
            lora_dropout=0.1,
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./gloss_to_text_t5"):
        # Use DataCollatorForSeq2Seq for T5
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=3,
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=2,
            eval_strategy='steps',
            eval_steps=100,
            save_steps=100,
            logging_steps=25,
            learning_rate=5e-4,
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_translation(self, gloss, max_new_tokens=50):
        input_text = f"translate english gloss to english text: {gloss}"

        inputs = self.tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # T5 generation
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=False,
                early_stopping=True,
                no_repeat_ngram_size=2
            )

        # Decode the generated tokens
        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        return generated_text.strip()

    def generate_translations(self, glosses, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_translation(g, max_new_tokens) for g in tqdm(glosses, desc="Generating translations")]

    def calculate_metrics(self, preds, refs):
        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess intensified-phoenix-14-t dataset
    pipeline = GlossToTextTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation
    samples = df[gloss_col].sample(3, random_state=42).tolist()
    outputs = pipeline.generate_translations(samples)
    for gloss, out in zip(samples, outputs):
        print(f"Gloss: {gloss} -> Text: {out}")

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_glosses = val_df[gloss_col].tolist()
    val_refs    = val_df[text_col].tolist()
    val_preds   = pipeline.generate_translations(val_glosses)
    metrics = pipeline.calculate_metrics(val_preds, val_refs)
    print("Validation Metrics:")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")
```

```python
#llama text to gloss
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling,
    BitsAndBytesConfig
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import gc

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class TextToGlossTranslationPipeline:
    def __init__(self, model_name="meta-llama/Meta-Llama-3-8B"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)
        if not gloss_col or not text_col:
            raise ValueError("Dataset must have columns for gloss and text.")

        # Drop nulls and ensure string type
        df = df.dropna(subset=[gloss_col, text_col]).astype({gloss_col: str, text_col: str})

        # Limit size for memory efficiency
        if len(df) > 2000:
            df = df.sample(2000, random_state=42).reset_index(drop=True)
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_prompt(text, gloss):
            return (
                "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n\n"
                "You are an expert translator who converts natural English text to English gloss notation. Follow these critical rules:\n"
                "NUMERICAL ACCURACY:\n"
                "- Preserve ALL numbers exactly as they appear in the text\n"
                "- Maintain original numerical formatting (percentages, decimals, years)\n"
                "- Never change, approximate, or substitute numerical values\n"
                "GLOSS NOTATION RULES:\n"
                "- Use X-WE for \"we\", X-I for \"I\", X-YOU for \"you\", X-HE for \"he\", X-SHE for \"she\"\n"
                "- Use X-POSS for possessive \"'s\"\n"
                "- Use DESC-[word] for adjectives/adverbs (e.g., \"important\" → \"DESC-IMPORTANT\")\n"
                "- Remove articles (a, an, the) unless semantically crucial\n"
                "- Convert verbs to base form unless tense is critical\n"
                "- Use hyphens to connect multi-word concepts\n"
                "FORMATTING REQUIREMENTS:\n"
                "- Use ALL CAPS for gloss words\n"
                "- Preserve percentage format exactly (e.g., \"10%\" → \"10%\")\n"
                "- Maintain word order that captures English structure\n"
                "- Remove auxiliary verbs unless they carry meaning\n"
                "LINGUISTIC PRECISION:\n"
                "- Focus on semantic content over grammatical markers\n"
                "- Preserve proper nouns and technical terms\n"
                "- Maintain plural markers when semantically important\n"
                "Convert this English text into gloss notation:\n\n"
                f"Text: {text}\n"
                f"Reference: {gloss}"
            )

        df['training_text'] = df.apply(lambda r: create_prompt(r[text_col], r[gloss_col]), axis=1)
        train_df, val_df = train_test_split(df[['training_text']], test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        if torch.cuda.is_available(): torch.cuda.empty_cache(); gc.collect()

        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        bnb_config = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_use_double_quant=True,
            bnb_4bit_quant_type="nf4",
            bnb_4bit_compute_dtype=torch.bfloat16
        )
        base_model = AutoModelForCausalLM.from_pretrained(
            self.model_name,
            quantization_config=bnb_config,
            device_map='auto',
            torch_dtype=torch.bfloat16,
            low_cpu_mem_usage=True,
            trust_remote_code=True
        )
        self.model = prepare_model_for_kbit_training(base_model)

        lora_cfg = LoraConfig(
            r=16,
            lora_alpha=32,
            target_modules=["q_proj","k_proj","v_proj","o_proj","gate_proj","up_proj","down_proj"],
            lora_dropout=0.1,
            bias="none",
            task_type=TaskType.CAUSAL_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(ex):
            tok = self.tokenizer(
                ex['training_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )
            tok['labels'] = tok['input_ids'].copy()
            return tok
        return ds.map(tokenize_fn, batched=True, remove_columns=['training_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./text_to_gloss_qlora"):
        if torch.cuda.is_available(): torch.cuda.empty_cache(); gc.collect()

        data_collator = DataCollatorForLanguageModeling(
            tokenizer=self.tokenizer,
            mlm=False,
            pad_to_multiple_of=8
        )
        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=3,
            per_device_train_batch_size=2,
            per_device_eval_batch_size=2,
            gradient_accumulation_steps=4,
            eval_strategy='steps',
            eval_steps=100,
            save_steps=100,
            logging_steps=25,
            learning_rate=2e-4,
            bf16=True,
            save_total_limit=3,
            remove_unused_columns=True,
            gradient_checkpointing=True,
            prediction_loss_only=True
        )
        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_translation(self, text, max_new_tokens=50):
        prompt = (
            "You are an expert translator from English text to gloss notation. \n"
            "CRITICAL REQUIREMENTS:\n"
            "1. PRESERVE ALL NUMBERS EXACTLY - never change any numerical values\n"
            "2. Use ALL CAPS for gloss words\n"
            "3. Convert pronouns: we→X-WE, I→X-I, you→X-YOU, he→X-HE, she→X-SHE\n"
            "4. Convert possessives: 's→X-POSS\n"
            "5. Convert adjectives/adverbs: use DESC- prefix (e.g., important→DESC-IMPORTANT)\n"
            "6. Remove articles (a, an, the) unless semantically crucial\n"
            "7. Use base verb forms unless tense is critical\n"
            "8. Connect multi-word concepts with hyphens\n"
            "9. Remove auxiliary verbs unless they carry meaning\n"
            "10. Maintain word order that captures English structure\n"
            "Translate the following text into gloss notation. Output ONLY the gloss:\n\n"
            f"Text: {text}\n"
            "Gloss:"
        )
        inputs = self.tokenizer(
            prompt,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        out = self.peft_model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            num_return_sequences=1,
            do_sample=False,
            early_stopping=True,
            eos_token_id=self.tokenizer.eos_token_id,
            pad_token_id=self.tokenizer.eos_token_id,
            no_repeat_ngram_size=2
        )
        raw = self.tokenizer.decode(out[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)
        first = raw.split('\n')[0].strip()
        return first

    def generate_translations(self, texts, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_translation(t, max_new_tokens) for t in tqdm(texts, desc="Generating glosses")]

    def calculate_bleu_score(self, preds, refs):
        res = self.bleu_metric.compute(predictions=preds, references=[[r] for r in refs])
        p = res.get('precisions', [0,0,0,0])
        return {'bleu':res['bleu'], 'bleu_1':p[0], 'bleu_2':p[1], 'bleu_3':p[2], 'bleu_4':p[3]}

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12
    pipeline = TextToGlossTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation
    samples = df[text_col].sample(3, random_state=42).tolist()
    outputs = pipeline.generate_translations(samples)
    for text, out in zip(samples, outputs):
        print(f"Text: {text} -> Gloss: {out}")

    # 5. Validation BLEU
    print("\nComputing BLEU on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_texts = val_df[text_col].tolist()
    val_refs  = val_df[gloss_col].tolist()
    val_preds = pipeline.generate_translations(val_texts)
    bleu_scores = pipeline.calculate_bleu_score(val_preds, val_refs)
    print("Validation BLEU Scores:", bleu_scores)
```

```python
#text to gloss with t5 - improved with preprocessing
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer
import re
import json
import spacy
from collections import defaultdict

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class TextPreprocessor:
    def __init__(self):
        # Load spaCy model for NLP processing
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except OSError:
            print("Warning: spaCy English model not found. Some preprocessing features will be limited.")
            self.nlp = None

        # Gender mapping for common titles
        self.gender_mapping = {
            'mr': 'mr', 'mrs': 'mrs', 'ms': 'ms', 'madam': 'madam',
            'president': 'president', 'commissioner': 'commissioner',
            'baroness': 'baroness', 'chairman': 'chairman', 'chairwoman': 'chairwoman'
        }

        # Common EU institutional terms
        self.institutional_terms = {
            'parliament', 'commission', 'council', 'european', 'union', 'eu',
            'committee', 'directive', 'regulation', 'treaty', 'member', 'state'
        }

        # Number patterns
        self.number_patterns = {
            'percentage': re.compile(r'(\d+)\s*%'),
            'year': re.compile(r'\b(19|20)\d{2}\b'),
            'decimal': re.compile(r'\d+\.\d+'),
            'currency': re.compile(r'EUR\s*\d+(?:\.\d+)?(?:\s*(?:billion|million))?', re.IGNORECASE)
        }

    def normalize_text_for_gloss(self, text):
        """Normalize English text for better gloss generation"""
        if not isinstance(text, str):
            return ""

        # Remove BOM and clean whitespace
        text = text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to lowercase for processing
        text = text.lower()

        # Expand contractions for clearer gloss representation
        contractions = {
            "can't": "cannot",
            "won't": "will not",
            "don't": "do not",
            "doesn't": "does not",
            "didn't": "did not",
            "isn't": "is not",
            "aren't": "are not",
            "wasn't": "was not",
            "weren't": "were not",
            "haven't": "have not",
            "hasn't": "has not",
            "hadn't": "had not",
            "shouldn't": "should not",
            "wouldn't": "would not",
            "couldn't": "could not",
            "mustn't": "must not",
            "needn't": "need not",
            "i'm": "i am",
            "you're": "you are",
            "he's": "he is",
            "she's": "she is",
            "it's": "it is",
            "we're": "we are",
            "they're": "they are",
            "i've": "i have",
            "you've": "you have",
            "we've": "we have",
            "they've": "they have",
            "i'll": "i will",
            "you'll": "you will",
            "he'll": "he will",
            "she'll": "she will",
            "we'll": "we will",
            "they'll": "they will"
        }

        for contraction, expansion in contractions.items():
            text = re.sub(r'\b' + re.escape(contraction) + r'\b', expansion, text, flags=re.IGNORECASE)

        # Remove articles that may not be necessary in gloss
        # (but keep some for context)
        text = re.sub(r'\bthe\s+(?=parliament|commission|council|union|budget|report|debate|directive)\b', '', text)

        # Normalize punctuation
        text = re.sub(r'\s*,\s*', ' , ', text)
        text = re.sub(r'\s*\.\s*', ' . ', text)
        text = re.sub(r'\s*!\s*', ' !', text)
        text = re.sub(r'\s*\?\s*', ' ?', text)

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def normalize_numbers_and_dates(self, text):
        """Normalize number and date formats"""
        # Normalize percentages
        text = re.sub(r'(\d+)\s*%', r'\1 percent', text)

        # Normalize years
        text = re.sub(r'\b(\d{4})\b', lambda m: m.group(1), text)

        # Normalize currency
        text = re.sub(r'EUR\s*(\d+(?:\.\d+)?)', r'eur \1', text, flags=re.IGNORECASE)

        return text

    def fix_common_errors(self, text):
        """Fix common systematic errors"""
        # Fix spacing issues
        text = re.sub(r'(\w)([,.])', r'\1 \2', text)  # Add space before punctuation
        text = re.sub(r'\s+([,.!?])', r' \1', text)    # Ensure single space before punctuation

        # Fix common word boundaries
        text = re.sub(r'inthe\b', 'in the', text, flags=re.IGNORECASE)
        text = re.sub(r'onthe\b', 'on the', text, flags=re.IGNORECASE)
        text = re.sub(r'tothe\b', 'to the', text, flags=re.IGNORECASE)

        return text

    def process_text(self, text):
        """Main preprocessing pipeline for English text"""
        if not isinstance(text, str) or not text.strip():
            return ""

        text = self.normalize_text_for_gloss(text)

        # Skip further processing if text is empty after normalization
        if not text.strip():
            return ""

        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def process_gloss_target(self, gloss_text):
        """Preprocessing for target gloss text"""
        if not isinstance(gloss_text, str):
            return ""

        # Remove BOM and clean whitespace
        text = gloss_text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to uppercase for gloss consistency
        text = text.upper()

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

class GlossPostProcessor:
    def __init__(self):
        # Common gloss corrections
        self.gloss_corrections = {
            'COMMISSIONS': 'COMMISSION',
            'PARLIAMENTS': 'PARLIAMENT',
        }

        # ASL gloss patterns to normalize
        self.gloss_patterns = {
            'possessive': re.compile(r"'S\b"),
            'plural_marker': re.compile(r'\b(\w+)S\b'),
        }

    def fix_gloss_structure(self, gloss):
        """Fix ASL gloss structural elements"""
        words = gloss.split()
        corrected_words = []

        for word in words:
            # Handle possessives
            if "'S" in word:
                word = word.replace("'S", "-POSS")

            # Handle common corrections
            if word in self.gloss_corrections:
                word = self.gloss_corrections[word]

            corrected_words.append(word)

        return ' '.join(corrected_words)

    def normalize_gloss_format(self, gloss):
        """Normalize gloss formatting"""
        # Ensure uppercase
        gloss = gloss.upper()

        # Clean up spacing
        gloss = re.sub(r'\s+', ' ', gloss).strip()

        # Remove punctuation that doesn't belong in gloss
        gloss = re.sub(r'[,.]', '', gloss)

        return gloss

    def post_process(self, gloss):
        """Main post-processing pipeline for generated gloss"""
        gloss = self.fix_gloss_structure(gloss)
        gloss = self.normalize_gloss_format(gloss)

        # Final cleanup
        gloss = re.sub(r'\s+', ' ', gloss).strip()

        return gloss

class TextToGlossTranslationPipeline:
    def __init__(self, model_name="t5-base"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

        # Initialize preprocessors (note the direction change)
        self.text_preprocessor = TextPreprocessor()
        self.gloss_post_processor = GlossPostProcessor()

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        print(f"Found columns: gloss='{gloss_col}', text='{text_col}'")
        print(f"Original dataset size: {len(df)}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])
        print(f"After dropping nulls: {len(df)}")

        # Create copies to preserve original data
        df = df.copy()
        original_gloss = df[gloss_col].copy()
        original_text = df[text_col].copy()

        # Apply preprocessing - note the direction change
        print("Applying text preprocessing...")
        df[text_col] = df[text_col].apply(self.text_preprocessor.process_text)

        print("Applying target gloss preprocessing...")
        df[gloss_col] = df[gloss_col].apply(self.text_preprocessor.process_gloss_target)

        # Filter out very short or very long sequences
        valid_mask = (
            (df[gloss_col].str.len() > 10) &  # Remove very short glosses
            (df[text_col].str.len() > 10) &   # Remove very short texts
            (df[gloss_col].str.len() < 500) & # Remove very long glosses
            (df[text_col].str.len() < 500)    # Remove very long texts
        )

        df = df[valid_mask].reset_index(drop=True)
        original_gloss = original_gloss[valid_mask].reset_index(drop=True)
        original_text = original_text[valid_mask].reset_index(drop=True)

        print(f"After length filtering: {len(df)}")

        # Limit size for memory efficiency
        if len(df) > 2000:
            sample_indices = df.sample(2000, random_state=42).index
            df = df.loc[sample_indices].reset_index(drop=True)
            original_gloss = original_gloss.loc[sample_indices].reset_index(drop=True)
            original_text = original_text.loc[sample_indices].reset_index(drop=True)

        # Add original columns for debugging
        df['original_gloss'] = original_gloss
        df['original_text'] = original_text

        print(f"Final dataset size: {len(df)} samples")
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(text, gloss):
            # T5 expects task prefix for sequence-to-sequence tasks
            # NOTE: Direction changed - text is input, gloss is target
            input_text = f"translate english text to american sign language gloss: {text}"
            target_gloss = gloss
            return input_text, target_gloss

        # Create input and target columns (note the parameter order change)
        df[['input_text', 'target_gloss']] = df.apply(
            lambda r: pd.Series(create_input_target(r[text_col], r[gloss_col])), axis=1
        )

        # Data augmentation: create variations with different prompts
        augmented_data = []
        for _, row in df.iterrows():
            # Original
            augmented_data.append({
                'input_text': row['input_text'],
                'target_text': row['target_gloss']  # Keep as target_text for compatibility
            })

            # Alternative prompt
            augmented_data.append({
                'input_text': f"convert text to gloss: {row[text_col]}",
                'target_text': row['target_gloss']
            })

        augmented_df = pd.DataFrame(augmented_data)
        train_df, val_df = train_test_split(augmented_df, test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        # Load T5 tokenizer and model
        self.tokenizer = T5Tokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # Load T5 model
        self.model = T5ForConditionalGeneration.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # Enhanced LoRA config for T5
        lora_cfg = LoraConfig(
            r=32,  # Increased rank for better performance
            lora_alpha=64,
            target_modules=["q", "v", "k", "o", "wi", "wo"],
            lora_dropout=0.05,  # Reduced dropout
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./text_to_gloss_t5"):
        # Use DataCollatorForSeq2Seq for T5
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=5,  # Increased epochs
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=4,  # Increased for better gradients
            eval_strategy='steps',
            eval_steps=50,
            save_steps=50,
            logging_steps=25,
            learning_rate=3e-4,  # Slightly reduced learning rate
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True,
            warmup_steps=100,  # Added warmup
            weight_decay=0.01  # Added weight decay
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_gloss(self, text, max_new_tokens=50):
        # Preprocess the text
        processed_text = self.text_preprocessor.process_text(text)
        input_text = f"translate english text to american sign language gloss: {processed_text}"

        inputs = self.tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # T5 generation with improved parameters
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=True,  # Enable sampling
                temperature=0.7,  # Add temperature for diversity
                top_p=0.9,       # Add nucleus sampling
                early_stopping=True,
                no_repeat_ngram_size=3,  # Increased to avoid repetition
                length_penalty=1.0
            )

        # Decode the generated tokens
        generated_gloss = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Apply post-processing
        processed_gloss = self.gloss_post_processor.post_process(generated_gloss.strip())

        return processed_gloss

    def generate_glosses(self, texts, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_gloss(t, max_new_tokens) for t in tqdm(texts, desc="Generating glosses")]

    def calculate_metrics(self, preds, refs):
        # Apply post-processing to predictions
        processed_preds = [self.gloss_post_processor.post_process(pred) for pred in preds]

        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=processed_preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=processed_preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=processed_preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(processed_preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = TextToGlossTranslationPipeline()
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # Show some preprocessing examples with original data
    print("\nPreprocessing examples:")
    for i in range(min(3, len(df))):
        print(f"Original text: {df.iloc[i]['original_text']}")
        print(f"Processed text: {df.iloc[i][text_col]}")
        print(f"Original gloss: {df.iloc[i]['original_gloss']}")
        print(f"Processed gloss: {df.iloc[i][gloss_col]}")
        print("-" * 50)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    print(f"Training samples: {len(tok_ds['train'])}")
    print(f"Validation samples: {len(tok_ds['validation'])}")

    # 3. Train
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation (now text to gloss)
    samples = df[text_col].sample(5, random_state=42).tolist()
    outputs = pipeline.generate_glosses(samples)
    print("\nSample translations:")
    for text, out in zip(samples, outputs):
        print(f"Text: {text}")
        print(f"Generated Gloss: {out}")
        print("-" * 50)

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_texts = val_df[text_col].tolist()  # Input is now text
    val_refs = val_df[gloss_col].tolist()  # Reference is now gloss
    val_preds = pipeline.generate_glosses(val_texts)  # Generate glosses from text
    metrics = pipeline.calculate_metrics(val_preds, val_refs)

    print("Validation Metrics:")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")

    # Show some prediction vs reference examples with original context
    print("\nSample predictions vs references:")
    for i in range(min(5, len(val_preds))):
        # Find corresponding original data
        original_idx = val_df.index[i]
        print(f"Original text: {df.loc[original_idx, 'original_text']}")
        print(f"Processed text: {val_texts[i]}")
        print(f"Reference gloss: {val_refs[i]}")
        print(f"Predicted gloss: {val_preds[i]}")
        print("-" * 50)
```

```python
#text to gloss with flan-t5 - improved with preprocessing
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer
import re
import json
import spacy
from collections import defaultdict

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class TextPreprocessor:
    def __init__(self):
        # Load spaCy model for NLP processing
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except OSError:
            print("Warning: spaCy English model not found. Some preprocessing features will be limited.")
            self.nlp = None

        # Gender mapping for common titles
        self.gender_mapping = {
            'mr': 'mr', 'mrs': 'mrs', 'ms': 'ms', 'madam': 'madam',
            'president': 'president', 'commissioner': 'commissioner',
            'baroness': 'baroness', 'chairman': 'chairman', 'chairwoman': 'chairwoman'
        }

        # Common EU institutional terms
        self.institutional_terms = {
            'parliament', 'commission', 'council', 'european', 'union', 'eu',
            'committee', 'directive', 'regulation', 'treaty', 'member', 'state'
        }

        # Number patterns
        self.number_patterns = {
            'percentage': re.compile(r'(\d+)\s*%'),
            'year': re.compile(r'\b(19|20)\d{2}\b'),
            'decimal': re.compile(r'\d+\.\d+'),
            'currency': re.compile(r'EUR\s*\d+(?:\.\d+)?(?:\s*(?:billion|million))?', re.IGNORECASE)
        }

    def normalize_text_for_gloss(self, text):
        """Normalize English text for better gloss generation"""
        if not isinstance(text, str):
            return ""

        # Remove BOM and clean whitespace
        text = text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to lowercase for processing
        text = text.lower()

        # Expand contractions for clearer gloss representation
        contractions = {
            "can't": "cannot",
            "won't": "will not",
            "don't": "do not",
            "doesn't": "does not",
            "didn't": "did not",
            "isn't": "is not",
            "aren't": "are not",
            "wasn't": "was not",
            "weren't": "were not",
            "haven't": "have not",
            "hasn't": "has not",
            "hadn't": "had not",
            "shouldn't": "should not",
            "wouldn't": "would not",
            "couldn't": "could not",
            "mustn't": "must not",
            "needn't": "need not",
            "i'm": "i am",
            "you're": "you are",
            "he's": "he is",
            "she's": "she is",
            "it's": "it is",
            "we're": "we are",
            "they're": "they are",
            "i've": "i have",
            "you've": "you have",
            "we've": "we have",
            "they've": "they have",
            "i'll": "i will",
            "you'll": "you will",
            "he'll": "he will",
            "she'll": "she will",
            "we'll": "we will",
            "they'll": "they will"
        }

        for contraction, expansion in contractions.items():
            text = re.sub(r'\b' + re.escape(contraction) + r'\b', expansion, text, flags=re.IGNORECASE)

        # Remove articles that may not be necessary in gloss
        # (but keep some for context)
        text = re.sub(r'\bthe\s+(?=parliament|commission|council|union|budget|report|debate|directive)\b', '', text)

        # Normalize punctuation
        text = re.sub(r'\s*,\s*', ' , ', text)
        text = re.sub(r'\s*\.\s*', ' . ', text)
        text = re.sub(r'\s*!\s*', ' !', text)
        text = re.sub(r'\s*\?\s*', ' ?', text)

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def normalize_numbers_and_dates(self, text):
        """Normalize number and date formats"""
        # Normalize percentages
        text = re.sub(r'(\d+)\s*%', r'\1 percent', text)

        # Normalize years
        text = re.sub(r'\b(\d{4})\b', lambda m: m.group(1), text)

        # Normalize currency
        text = re.sub(r'EUR\s*(\d+(?:\.\d+)?)', r'eur \1', text, flags=re.IGNORECASE)

        return text

    def fix_common_errors(self, text):
        """Fix common systematic errors"""
        # Fix spacing issues
        text = re.sub(r'(\w)([,.])', r'\1 \2', text)  # Add space before punctuation
        text = re.sub(r'\s+([,.!?])', r' \1', text)    # Ensure single space before punctuation

        # Fix common word boundaries
        text = re.sub(r'inthe\b', 'in the', text, flags=re.IGNORECASE)
        text = re.sub(r'onthe\b', 'on the', text, flags=re.IGNORECASE)
        text = re.sub(r'tothe\b', 'to the', text, flags=re.IGNORECASE)

        return text

    def process_text(self, text):
        """Main preprocessing pipeline for English text"""
        if not isinstance(text, str) or not text.strip():
            return ""

        text = self.normalize_text_for_gloss(text)

        # Skip further processing if text is empty after normalization
        if not text.strip():
            return ""

        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def process_gloss_target(self, gloss_text):
        """Preprocessing for target gloss text"""
        if not isinstance(gloss_text, str):
            return ""

        # Remove BOM and clean whitespace
        text = gloss_text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to uppercase for gloss consistency
        text = text.upper()

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

class GlossPostProcessor:
    def __init__(self):
        # Common gloss corrections
        self.gloss_corrections = {
            'COMMISSIONS': 'COMMISSION',
            'PARLIAMENTS': 'PARLIAMENT',
        }

        # ASL gloss patterns to normalize
        self.gloss_patterns = {
            'possessive': re.compile(r"'S\b"),
            'plural_marker': re.compile(r'\b(\w+)S\b'),
        }

    def fix_gloss_structure(self, gloss):
        """Fix ASL gloss structural elements"""
        words = gloss.split()
        corrected_words = []

        for word in words:
            # Handle possessives
            if "'S" in word:
                word = word.replace("'S", "-POSS")

            # Handle common corrections
            if word in self.gloss_corrections:
                word = self.gloss_corrections[word]

            corrected_words.append(word)

        return ' '.join(corrected_words)

    def normalize_gloss_format(self, gloss):
        """Normalize gloss formatting"""
        # Ensure uppercase
        gloss = gloss.upper()

        # Clean up spacing
        gloss = re.sub(r'\s+', ' ', gloss).strip()

        # Remove punctuation that doesn't belong in gloss
        gloss = re.sub(r'[,.]', '', gloss)

        return gloss

    def post_process(self, gloss):
        """Main post-processing pipeline for generated gloss"""
        gloss = self.fix_gloss_structure(gloss)
        gloss = self.normalize_gloss_format(gloss)

        # Final cleanup
        gloss = re.sub(r'\s+', ' ', gloss).strip()

        return gloss

class TextToGlossTranslationPipeline:
    def __init__(self, model_name="google/flan-t5-base"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

        # Initialize preprocessors (note the direction change)
        self.text_preprocessor = TextPreprocessor()
        self.gloss_post_processor = GlossPostProcessor()

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        print(f"Found columns: gloss='{gloss_col}', text='{text_col}'")
        print(f"Original dataset size: {len(df)}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])
        print(f"After dropping nulls: {len(df)}")

        # Create copies to preserve original data
        df = df.copy()
        original_gloss = df[gloss_col].copy()
        original_text = df[text_col].copy()

        # Apply preprocessing - note the direction change
        print("Applying text preprocessing...")
        df[text_col] = df[text_col].apply(self.text_preprocessor.process_text)

        print("Applying target gloss preprocessing...")
        df[gloss_col] = df[gloss_col].apply(self.text_preprocessor.process_gloss_target)

        # Filter out very short or very long sequences
        valid_mask = (
            (df[gloss_col].str.len() > 10) &  # Remove very short glosses
            (df[text_col].str.len() > 10) &   # Remove very short texts
            (df[gloss_col].str.len() < 500) & # Remove very long glosses
            (df[text_col].str.len() < 500)    # Remove very long texts
        )

        df = df[valid_mask].reset_index(drop=True)
        original_gloss = original_gloss[valid_mask].reset_index(drop=True)
        original_text = original_text[valid_mask].reset_index(drop=True)

        print(f"After length filtering: {len(df)}")

        # Limit size for memory efficiency
        if len(df) > 2000:
            sample_indices = df.sample(2000, random_state=42).index
            df = df.loc[sample_indices].reset_index(drop=True)
            original_gloss = original_gloss.loc[sample_indices].reset_index(drop=True)
            original_text = original_text.loc[sample_indices].reset_index(drop=True)

        # Add original columns for debugging
        df['original_gloss'] = original_gloss
        df['original_text'] = original_text

        print(f"Final dataset size: {len(df)} samples")
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(text, gloss):
            # FLAN-T5 works better with more natural instruction prompts
            # Note: Direction changed - text is input, gloss is target
            input_text = f"Translate this English text into American Sign Language gloss notation: {text}"
            target_gloss = gloss
            return input_text, target_gloss

        # Create input and target columns (note the parameter order change)
        df[['input_text', 'target_gloss']] = df.apply(
            lambda r: pd.Series(create_input_target(r[text_col], r[gloss_col])), axis=1
        )

        # Data augmentation: create variations with different prompts
        # FLAN-T5 benefits from diverse instruction formats
        augmented_data = []
        for _, row in df.iterrows():
            # Original instruction format
            augmented_data.append({
                'input_text': row['input_text'],
                'target_text': row['target_gloss']  # Keep as target_text for compatibility
            })

            # Alternative instruction format
            augmented_data.append({
                'input_text': f"Convert the following English sentence to ASL gloss: {row[text_col]}",
                'target_text': row['target_gloss']
            })

            # Another variation
            augmented_data.append({
                'input_text': f"How would you express this in American Sign Language gloss notation? {row[text_col]}",
                'target_text': row['target_gloss']
            })

        augmented_df = pd.DataFrame(augmented_data)
        train_df, val_df = train_test_split(augmented_df, test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        # Load FLAN-T5 tokenizer and model
        self.tokenizer = T5Tokenizer.from_pretrained(self.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # Load FLAN-T5 model
        self.model = T5ForConditionalGeneration.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # Enhanced LoRA config for FLAN-T5
        # FLAN-T5 typically benefits from slightly different LoRA parameters
        lora_cfg = LoraConfig(
            r=16,  # Slightly lower rank as FLAN-T5 is already well-tuned
            lora_alpha=32,
            target_modules=["q", "v", "k", "o", "wi", "wo"],
            lora_dropout=0.1,  # Slightly higher dropout for FLAN-T5
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./text_to_gloss_flan_t5"):
        # Use DataCollatorForSeq2Seq for FLAN-T5
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        # FLAN-T5 specific training arguments
        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=3,  # FLAN-T5 typically needs fewer epochs due to pre-training
            per_device_train_batch_size=4,
            per_device_eval_batch_size=4,
            gradient_accumulation_steps=4,
            eval_strategy='steps',
            eval_steps=50,
            save_steps=50,
            logging_steps=25,
            learning_rate=1e-4,  # Lower learning rate for FLAN-T5
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True,
            warmup_steps=50,  # Reduced warmup for FLAN-T5
            weight_decay=0.01,
            dataloader_drop_last=True,  # Help with consistent batch sizes
            load_best_model_at_end=True,  # Load best model at end
            metric_for_best_model="eval_loss",
            greater_is_better=False
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_gloss(self, text, max_new_tokens=50):
        # Preprocess the text
        processed_text = self.text_preprocessor.process_text(text)
        # Use natural instruction format for FLAN-T5
        input_text = f"Translate this English text into American Sign Language gloss notation: {processed_text}"

        inputs = self.tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # FLAN-T5 generation with optimized parameters
        with torch.no_grad():
            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=True,  # Enable sampling
                temperature=0.8,  # Slightly higher temperature for FLAN-T5
                top_p=0.95,      # Higher top_p for FLAN-T5
                early_stopping=True,
                no_repeat_ngram_size=2,  # Prevent repetition
                length_penalty=1.2,  # Encourage longer outputs if needed
                pad_token_id=self.tokenizer.pad_token_id,
                eos_token_id=self.tokenizer.eos_token_id
            )

        # Decode the generated tokens
        generated_gloss = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Apply post-processing
        processed_gloss = self.gloss_post_processor.post_process(generated_gloss.strip())

        return processed_gloss

    def generate_glosses(self, texts, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_gloss(t, max_new_tokens) for t in tqdm(texts, desc="Generating glosses")]

    def calculate_metrics(self, preds, refs):
        # Apply post-processing to predictions
        processed_preds = [self.gloss_post_processor.post_process(pred) for pred in preds]

        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=processed_preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=processed_preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=processed_preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(processed_preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = TextToGlossTranslationPipeline()  # Now uses FLAN-T5 by default
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # Show some preprocessing examples with original data
    print("\nPreprocessing examples:")
    for i in range(min(3, len(df))):
        print(f"Original text: {df.iloc[i]['original_text']}")
        print(f"Processed text: {df.iloc[i][text_col]}")
        print(f"Original gloss: {df.iloc[i]['original_gloss']}")
        print(f"Processed gloss: {df.iloc[i][gloss_col]}")
        print("-" * 50)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    print(f"Training samples: {len(tok_ds['train'])}")
    print(f"Validation samples: {len(tok_ds['validation'])}")

    # 3. Train with FLAN-T5
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation (now text to gloss with FLAN-T5)
    samples = df[text_col].sample(5, random_state=42).tolist()
    outputs = pipeline.generate_glosses(samples)
    print("\nSample translations with FLAN-T5:")
    for text, out in zip(samples, outputs):
        print(f"Text: {text}")
        print(f"Generated Gloss: {out}")
        print("-" * 50)

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_texts = val_df[text_col].tolist()  # Input is now text
    val_refs = val_df[gloss_col].tolist()  # Reference is now gloss
    val_preds = pipeline.generate_glosses(val_texts)  # Generate glosses from text
    metrics = pipeline.calculate_metrics(val_preds, val_refs)

    print("Validation Metrics (FLAN-T5):")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")

    # Show some prediction vs reference examples with original context
    print("\nSample predictions vs references (FLAN-T5):")
    for i in range(min(5, len(val_preds))):
        # Find corresponding original data
        original_idx = val_df.index[i]
        print(f"Original text: {df.loc[original_idx, 'original_text']}")
        print(f"Processed text: {val_texts[i]}")
        print(f"Reference gloss: {val_refs[i]}")
        print(f"Predicted gloss: {val_preds[i]}")
        print("-" * 50)
```

```python
#text to gloss with mbart - improved with preprocessing
import torch
import pandas as pd
from datasets import Dataset, DatasetDict, load_dataset
from transformers import (
    MBartTokenizer,
    MBartForConditionalGeneration,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
import evaluate
from tqdm import tqdm
import warnings
import os
import jiwer
import re
import json
import spacy
from collections import defaultdict

# Disable external logging
os.environ["WANDB_DISABLED"] = "true"
warnings.filterwarnings("ignore")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class TextPreprocessor:
    def __init__(self):
        # Load spaCy model for NLP processing
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except OSError:
            print("Warning: spaCy English model not found. Some preprocessing features will be limited.")
            self.nlp = None

        # Gender mapping for common titles
        self.gender_mapping = {
            'mr': 'mr', 'mrs': 'mrs', 'ms': 'ms', 'madam': 'madam',
            'president': 'president', 'commissioner': 'commissioner',
            'baroness': 'baroness', 'chairman': 'chairman', 'chairwoman': 'chairwoman'
        }

        # Common EU institutional terms
        self.institutional_terms = {
            'parliament', 'commission', 'council', 'european', 'union', 'eu',
            'committee', 'directive', 'regulation', 'treaty', 'member', 'state'
        }

        # Number patterns
        self.number_patterns = {
            'percentage': re.compile(r'(\d+)\s*%'),
            'year': re.compile(r'\b(19|20)\d{2}\b'),
            'decimal': re.compile(r'\d+\.\d+'),
            'currency': re.compile(r'EUR\s*\d+(?:\.\d+)?(?:\s*(?:billion|million))?', re.IGNORECASE)
        }

    def normalize_text_for_gloss(self, text):
        """Normalize English text for better gloss generation"""
        if not isinstance(text, str):
            return ""

        # Remove BOM and clean whitespace
        text = text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to lowercase for processing
        text = text.lower()

        # Expand contractions for clearer gloss representation
        contractions = {
            "can't": "cannot",
            "won't": "will not",
            "don't": "do not",
            "doesn't": "does not",
            "didn't": "did not",
            "isn't": "is not",
            "aren't": "are not",
            "wasn't": "was not",
            "weren't": "were not",
            "haven't": "have not",
            "hasn't": "has not",
            "hadn't": "had not",
            "shouldn't": "should not",
            "wouldn't": "would not",
            "couldn't": "could not",
            "mustn't": "must not",
            "needn't": "need not",
            "i'm": "i am",
            "you're": "you are",
            "he's": "he is",
            "she's": "she is",
            "it's": "it is",
            "we're": "we are",
            "they're": "they are",
            "i've": "i have",
            "you've": "you have",
            "we've": "we have",
            "they've": "they have",
            "i'll": "i will",
            "you'll": "you will",
            "he'll": "he will",
            "she'll": "she will",
            "we'll": "we will",
            "they'll": "they will"
        }

        for contraction, expansion in contractions.items():
            text = re.sub(r'\b' + re.escape(contraction) + r'\b', expansion, text, flags=re.IGNORECASE)

        # Remove articles that may not be necessary in gloss
        # (but keep some for context)
        text = re.sub(r'\bthe\s+(?=parliament|commission|council|union|budget|report|debate|directive)\b', '', text)

        # Normalize punctuation
        text = re.sub(r'\s*,\s*', ' , ', text)
        text = re.sub(r'\s*\.\s*', ' . ', text)
        text = re.sub(r'\s*!\s*', ' !', text)
        text = re.sub(r'\s*\?\s*', ' ?', text)

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def normalize_numbers_and_dates(self, text):
        """Normalize number and date formats"""
        # Normalize percentages
        text = re.sub(r'(\d+)\s*%', r'\1 percent', text)

        # Normalize years
        text = re.sub(r'\b(\d{4})\b', lambda m: m.group(1), text)

        # Normalize currency
        text = re.sub(r'EUR\s*(\d+(?:\.\d+)?)', r'eur \1', text, flags=re.IGNORECASE)

        return text

    def fix_common_errors(self, text):
        """Fix common systematic errors"""
        # Fix spacing issues
        text = re.sub(r'(\w)([,.])', r'\1 \2', text)  # Add space before punctuation
        text = re.sub(r'\s+([,.!?])', r' \1', text)    # Ensure single space before punctuation

        # Fix common word boundaries
        text = re.sub(r'inthe\b', 'in the', text, flags=re.IGNORECASE)
        text = re.sub(r'onthe\b', 'on the', text, flags=re.IGNORECASE)
        text = re.sub(r'tothe\b', 'to the', text, flags=re.IGNORECASE)

        return text

    def process_text(self, text):
        """Main preprocessing pipeline for English text"""
        if not isinstance(text, str) or not text.strip():
            return ""

        text = self.normalize_text_for_gloss(text)

        # Skip further processing if text is empty after normalization
        if not text.strip():
            return ""

        text = self.normalize_numbers_and_dates(text)
        text = self.fix_common_errors(text)

        # Final cleanup
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def process_gloss_target(self, gloss_text):
        """Preprocessing for target gloss text"""
        if not isinstance(gloss_text, str):
            return ""

        # Remove BOM and clean whitespace
        text = gloss_text.strip().replace('\ufeff', '').replace('\xa0', ' ')

        # Convert to uppercase for gloss consistency
        text = text.upper()

        # Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

class GlossPostProcessor:
    def __init__(self):
        # Common gloss corrections
        self.gloss_corrections = {
            'COMMISSIONS': 'COMMISSION',
            'PARLIAMENTS': 'PARLIAMENT',
        }

        # ASL gloss patterns to normalize
        self.gloss_patterns = {
            'possessive': re.compile(r"'S\b"),
            'plural_marker': re.compile(r'\b(\w+)S\b'),
        }

    def fix_gloss_structure(self, gloss):
        """Fix ASL gloss structural elements"""
        words = gloss.split()
        corrected_words = []

        for word in words:
            # Handle possessives
            if "'S" in word:
                word = word.replace("'S", "-POSS")

            # Handle common corrections
            if word in self.gloss_corrections:
                word = self.gloss_corrections[word]

            corrected_words.append(word)

        return ' '.join(corrected_words)

    def normalize_gloss_format(self, gloss):
        """Normalize gloss formatting"""
        # Ensure uppercase
        gloss = gloss.upper()

        # Clean up spacing
        gloss = re.sub(r'\s+', ' ', gloss).strip()

        # Remove punctuation that doesn't belong in gloss
        gloss = re.sub(r'[,.]', '', gloss)

        return gloss

    def post_process(self, gloss):
        """Main post-processing pipeline for generated gloss"""
        gloss = self.fix_gloss_structure(gloss)
        gloss = self.normalize_gloss_format(gloss)

        # Final cleanup
        gloss = re.sub(r'\s+', ' ', gloss).strip()

        return gloss

class TextToGlossTranslationPipeline:
    def __init__(self, model_name="facebook/mbart-large-50-many-to-many-mmt"):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.peft_model = None
        self.bleu_metric = evaluate.load("bleu")
        self.rouge_metric = evaluate.load("rouge")
        self.meteor_metric = evaluate.load("meteor")

        # Initialize preprocessors
        self.text_preprocessor = TextPreprocessor()
        self.gloss_post_processor = GlossPostProcessor()

    def load_dataset(self, dataset_name="achrafothman/aslg_pc12"):
        """Load ASLG-PC12 dataset from HuggingFace"""
        return load_dataset(dataset_name)

    def preprocess_data(self, dataset):
        # Choose 'train' split if available, else first split
        split = dataset.get('train', next(iter(dataset.values())))
        df = split.to_pandas()

        # Identify columns dynamically for ASLG-PC12
        gloss_col = next((c for c in df.columns if 'gloss' in c.lower()), None)
        text_col  = next((c for c in df.columns if 'text'  in c.lower()), None)

        if not gloss_col or not text_col:
            raise ValueError(f"Dataset must have columns for gloss and text. Available columns: {df.columns.tolist()}")

        print(f"Found columns: gloss='{gloss_col}', text='{text_col}'")
        print(f"Original dataset size: {len(df)}")

        # Drop rows with null values in either column
        df = df.dropna(subset=[gloss_col, text_col])
        print(f"After dropping nulls: {len(df)}")

        # Create copies to preserve original data
        df = df.copy()
        original_gloss = df[gloss_col].copy()
        original_text = df[text_col].copy()

        # Apply preprocessing - note the direction change
        print("Applying text preprocessing...")
        df[text_col] = df[text_col].apply(self.text_preprocessor.process_text)

        print("Applying target gloss preprocessing...")
        df[gloss_col] = df[gloss_col].apply(self.text_preprocessor.process_gloss_target)

        # Filter out very short or very long sequences
        valid_mask = (
            (df[gloss_col].str.len() > 10) &  # Remove very short glosses
            (df[text_col].str.len() > 10) &   # Remove very short texts
            (df[gloss_col].str.len() < 500) & # Remove very long glosses
            (df[text_col].str.len() < 500)    # Remove very long texts
        )

        df = df[valid_mask].reset_index(drop=True)
        original_gloss = original_gloss[valid_mask].reset_index(drop=True)
        original_text = original_text[valid_mask].reset_index(drop=True)

        print(f"After length filtering: {len(df)}")

        # Limit size for memory efficiency
        if len(df) > 2000:
            sample_indices = df.sample(2000, random_state=42).index
            df = df.loc[sample_indices].reset_index(drop=True)
            original_gloss = original_gloss.loc[sample_indices].reset_index(drop=True)
            original_text = original_text.loc[sample_indices].reset_index(drop=True)

        # Add original columns for debugging
        df['original_gloss'] = original_gloss
        df['original_text'] = original_text

        print(f"Final dataset size: {len(df)} samples")
        return df, gloss_col, text_col

    def prepare_data_for_training(self, df, gloss_col, text_col, test_size=0.2):
        def create_input_target(text, gloss):
            # mBART doesn't need task prefixes like T5/FLAN-T5
            # Just provide the text directly
            input_text = text
            target_gloss = gloss
            return input_text, target_gloss

        # Create input and target columns
        df[['input_text', 'target_gloss']] = df.apply(
            lambda r: pd.Series(create_input_target(r[text_col], r[gloss_col])), axis=1
        )

        # Data augmentation: create variations with slight modifications
        # mBART benefits from clean, direct examples
        augmented_data = []
        for _, row in df.iterrows():
            # Original
            augmented_data.append({
                'input_text': row['input_text'],
                'target_text': row['target_gloss']  # Keep as target_text for compatibility
            })

            # Add a small variation (remove some articles)
            modified_text = re.sub(r'\ba\s+', '', row['input_text'])
            modified_text = re.sub(r'\ban\s+', '', modified_text)
            if modified_text.strip() and modified_text != row['input_text']:
                augmented_data.append({
                    'input_text': modified_text,
                    'target_text': row['target_gloss']
                })

        augmented_df = pd.DataFrame(augmented_data)
        train_df, val_df = train_test_split(augmented_df, test_size=test_size, random_state=42)

        train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
        val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
        return DatasetDict({'train': train_ds, 'validation': val_ds})

    def load_model_and_tokenizer(self):
        # Load mBART tokenizer and model
        self.tokenizer = MBartTokenizer.from_pretrained(self.model_name)

        # Set source and target languages
        # Using 'en_XX' for English source and creating a custom token for ASL gloss
        self.tokenizer.src_lang = "en_XX"
        self.tokenizer.tgt_lang = "en_XX"  # We'll use English tokens for gloss representation

        # Load mBART model
        self.model = MBartForConditionalGeneration.from_pretrained(
            self.model_name,
            torch_dtype=torch.float32 if device.type == 'cpu' else torch.bfloat16,
            device_map='auto' if torch.cuda.is_available() else None
        )

        # LoRA config for mBART
        lora_cfg = LoraConfig(
            r=16,  # Moderate rank for mBART
            lora_alpha=32,
            target_modules=["q_proj", "v_proj", "k_proj", "out_proj", "fc1", "fc2"],  # mBART specific modules
            lora_dropout=0.1,
            bias="none",
            task_type=TaskType.SEQ_2_SEQ_LM
        )
        self.peft_model = get_peft_model(self.model, lora_cfg)
        return self.tokenizer, self.peft_model

    def tokenize_data(self, ds, max_length=512):
        def tokenize_fn(examples):
            # Set source language for mBART
            self.tokenizer.src_lang = "en_XX"

            # Tokenize inputs
            model_inputs = self.tokenizer(
                examples['input_text'],
                truncation=True,
                padding='max_length',
                max_length=max_length
            )

            # Set target language for mBART
            self.tokenizer.tgt_lang = "en_XX"

            # Tokenize targets
            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    truncation=True,
                    padding='max_length',
                    max_length=max_length
                )

            model_inputs['labels'] = labels['input_ids']
            return model_inputs

        return ds.map(tokenize_fn, batched=True, remove_columns=['input_text', 'target_text'], batch_size=32)

    def train_model(self, tok_ds, output_dir="./text_to_gloss_mbart"):
        # Use DataCollatorForSeq2Seq for mBART
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.peft_model,
            pad_to_multiple_of=8
        )

        # mBART specific training arguments
        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=4,  # mBART typically needs moderate number of epochs
            per_device_train_batch_size=2,  # Smaller batch size due to mBART's size
            per_device_eval_batch_size=2,
            gradient_accumulation_steps=8,  # Increase to compensate for smaller batch size
            eval_strategy='steps',
            eval_steps=100,
            save_steps=100,
            logging_steps=50,
            learning_rate=5e-5,  # Conservative learning rate for mBART
            bf16=torch.cuda.is_available(),
            save_total_limit=3,
            remove_unused_columns=True,
            prediction_loss_only=True,
            warmup_steps=200,  # More warmup for mBART
            weight_decay=0.01,
            dataloader_drop_last=True,
            load_best_model_at_end=True,
            metric_for_best_model="eval_loss",
            greater_is_better=False,
            fp16=not torch.cuda.is_available() or not torch.cuda.get_device_capability()[0] >= 8,  # Use fp16 if no bf16 support
        )

        trainer = Trainer(
            model=self.peft_model,
            args=training_args,
            train_dataset=tok_ds['train'],
            eval_dataset=tok_ds['validation'],
            tokenizer=self.tokenizer,
            data_collator=data_collator
        )
        trainer.train()
        trainer.save_model()
        self.tokenizer.save_pretrained(output_dir)
        return trainer

    def generate_single_gloss(self, text, max_new_tokens=50):
        # Preprocess the text
        processed_text = self.text_preprocessor.process_text(text)

        # Set source language
        self.tokenizer.src_lang = "en_XX"

        inputs = self.tokenizer(
            processed_text,
            return_tensors="pt",
            truncation=True,
            max_length=256
        ).to(self.peft_model.device)

        # mBART generation parameters
        with torch.no_grad():
            # Set target language
            self.tokenizer.tgt_lang = "en_XX"

            outputs = self.peft_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                num_return_sequences=1,
                do_sample=True,
                temperature=0.7,  # Moderate temperature for mBART
                top_p=0.9,
                early_stopping=True,
                no_repeat_ngram_size=3,
                length_penalty=1.1,
                pad_token_id=self.tokenizer.pad_token_id,
                eos_token_id=self.tokenizer.eos_token_id,
                forced_bos_token_id=self.tokenizer.lang_code_to_id["en_XX"]  # Force English target
            )

        # Decode the generated tokens
        generated_gloss = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Apply post-processing
        processed_gloss = self.gloss_post_processor.post_process(generated_gloss.strip())

        return processed_gloss

    def generate_glosses(self, texts, max_new_tokens=50):
        self.peft_model.eval()
        return [self.generate_single_gloss(t, max_new_tokens) for t in tqdm(texts, desc="Generating glosses")]

    def calculate_metrics(self, preds, refs):
        # Apply post-processing to predictions
        processed_preds = [self.gloss_post_processor.post_process(pred) for pred in preds]

        # Calculate BLEU score
        bleu_res = self.bleu_metric.compute(predictions=processed_preds, references=[[r] for r in refs])
        bleu_p = bleu_res.get('precisions', [0,0,0,0])

        # Calculate ROUGE-L score
        rouge_res = self.rouge_metric.compute(predictions=processed_preds, references=refs)

        # Calculate METEOR score
        meteor_res = self.meteor_metric.compute(predictions=processed_preds, references=refs)

        # Calculate WER (Word Error Rate)
        wer_scores = []
        for pred, ref in zip(processed_preds, refs):
            try:
                wer_score = jiwer.wer(ref, pred)
                wer_scores.append(wer_score)
            except:
                # Handle edge cases where WER calculation fails
                wer_scores.append(1.0)

        avg_wer = sum(wer_scores) / len(wer_scores)

        return {
            'bleu': bleu_res['bleu'],
            'bleu_1': bleu_p[0],
            'bleu_2': bleu_p[1],
            'bleu_3': bleu_p[2],
            'bleu_4': bleu_p[3],
            'rouge_l': rouge_res['rougeL'],
            'meteor': meteor_res['meteor'],
            'wer': avg_wer
        }

if __name__ == "__main__":
    # 1. Load & preprocess ASLG-PC12 dataset
    pipeline = TextToGlossTranslationPipeline()  # Now uses mBART by default
    raw_ds = pipeline.load_dataset()
    df, gloss_col, text_col = pipeline.preprocess_data(raw_ds)

    # Show some preprocessing examples with original data
    print("\nPreprocessing examples:")
    for i in range(min(3, len(df))):
        print(f"Original text: {df.iloc[i]['original_text']}")
        print(f"Processed text: {df.iloc[i][text_col]}")
        print(f"Original gloss: {df.iloc[i]['original_gloss']}")
        print(f"Processed gloss: {df.iloc[i][gloss_col]}")
        print("-" * 50)

    # 2. Prepare & tokenize
    ds = pipeline.prepare_data_for_training(df, gloss_col, text_col)
    tokenizer, _ = pipeline.load_model_and_tokenizer()
    tok_ds = pipeline.tokenize_data(ds)

    print(f"Training samples: {len(tok_ds['train'])}")
    print(f"Validation samples: {len(tok_ds['validation'])}")

    # 3. Train with mBART
    trainer = pipeline.train_model(tok_ds)

    # 4. Sample generation (now text to gloss with mBART)
    samples = df[text_col].sample(5, random_state=42).tolist()
    outputs = pipeline.generate_glosses(samples)
    print("\nSample translations with mBART:")
    for text, out in zip(samples, outputs):
        print(f"Text: {text}")
        print(f"Generated Gloss: {out}")
        print("-" * 50)

    # 5. Validation metrics
    print("\nComputing metrics on validation set...")
    _, val_df = train_test_split(df, test_size=0.2, random_state=42)
    val_texts = val_df[text_col].tolist()  # Input is now text
    val_refs = val_df[gloss_col].tolist()  # Reference is now gloss
    val_preds = pipeline.generate_glosses(val_texts)  # Generate glosses from text
    metrics = pipeline.calculate_metrics(val_preds, val_refs)

    print("Validation Metrics (mBART):")
    print(f"  BLEU: {metrics['bleu']:.4f}")
    print(f"  BLEU-1: {metrics['bleu_1']:.4f}")
    print(f"  BLEU-2: {metrics['bleu_2']:.4f}")
    print(f"  BLEU-3: {metrics['bleu_3']:.4f}")
    print(f"  BLEU-4: {metrics['bleu_4']:.4f}")
    print(f"  ROUGE-L: {metrics['rouge_l']:.4f}")
    print(f"  METEOR: {metrics['meteor']:.4f}")
    print(f"  WER: {metrics['wer']:.4f}")

    # Show some prediction vs reference examples with original context
    print("\nSample predictions vs references (mBART):")
    for i in range(min(5, len(val_preds))):
        # Find corresponding original data
        original_idx = val_df.index[i]
        print(f"Original text: {df.loc[original_idx, 'original_text']}")
        print(f"Processed text: {val_texts[i]}")
        print(f"Reference gloss: {val_refs[i]}")
        print(f"Predicted gloss: {val_preds[i]}")
        print("-" * 50)
```

```python
#transformer signum gloss to text
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
import numpy as np
import math
from collections import Counter
import random
from typing import List, Tuple, Dict
import re

# For evaluation metrics
from nltk.translate.bleu_score import sentence_bleu, corpus_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score
from rouge_score import rouge_scorer
import nltk
import jiwer  # For WER calculation

# Download required NLTK data
try:
    nltk.download('wordnet', quiet=True)
    nltk.download('punkt', quiet=True)
except:
    pass

# ==================== Data Processing ====================

class SignumDataset(Dataset):
    """Dataset class for SIGNUM sign language translation"""

    def __init__(self, gloss_sentences, text_sentences, vocab_gloss=None, vocab_text=None,
                 max_length=100):
        self.gloss_sentences = gloss_sentences
        self.text_sentences = text_sentences
        self.max_length = max_length

        # Build vocabularies if not provided
        if vocab_gloss is None:
            self.vocab_gloss = self.build_vocab(gloss_sentences)
        else:
            self.vocab_gloss = vocab_gloss

        if vocab_text is None:
            self.vocab_text = self.build_vocab(text_sentences)
        else:
            self.vocab_text = vocab_text

        self.gloss_to_idx = {token: idx for idx, token in enumerate(self.vocab_gloss)}
        self.text_to_idx = {token: idx for idx, token in enumerate(self.vocab_text)}
        self.idx_to_text = {idx: token for token, idx in self.text_to_idx.items()}

    def build_vocab(self, sentences):
        """Build vocabulary from sentences"""
        vocab = ['<pad>', '<sos>', '<eos>', '<unk>']
        word_freq = Counter()

        for sentence in sentences:
            tokens = sentence.lower().split()
            word_freq.update(tokens)

        # Add words with frequency > 1
        for word, freq in word_freq.items():
            if freq > 1 and word not in vocab:
                vocab.append(word)

        return vocab

    def encode_sentence(self, sentence, vocab_to_idx, add_special_tokens=True):
        """Encode a sentence to indices"""
        tokens = sentence.lower().split()

        if add_special_tokens:
            tokens = ['<sos>'] + tokens + ['<eos>']

        indices = []
        for token in tokens[:self.max_length]:
            if token in vocab_to_idx:
                indices.append(vocab_to_idx[token])
            else:
                indices.append(vocab_to_idx['<unk>'])

        return torch.tensor(indices, dtype=torch.long)

    def __len__(self):
        return len(self.gloss_sentences)

    def __getitem__(self, idx):
        gloss = self.encode_sentence(self.gloss_sentences[idx], self.gloss_to_idx)
        text = self.encode_sentence(self.text_sentences[idx], self.text_to_idx)
        return gloss, text

# ==================== Transformer Model ====================

class PositionalEncoding(nn.Module):
    """Positional encoding for transformer"""

    def __init__(self, d_model, max_len=5000):
        super().__init__()

        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() *
                           (-math.log(10000.0) / d_model))

        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)

        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:x.size(0), :]

class TransformerSLT(nn.Module):
    """Transformer model for Sign Language Translation"""

    def __init__(self, src_vocab_size, tgt_vocab_size, d_model=512, nhead=8,
                 num_encoder_layers=6, num_decoder_layers=6, dim_feedforward=2048,
                 dropout=0.1, max_seq_length=100):
        super().__init__()

        self.d_model = d_model
        self.src_embedding = nn.Embedding(src_vocab_size, d_model)
        self.tgt_embedding = nn.Embedding(tgt_vocab_size, d_model)
        self.pos_encoder = PositionalEncoding(d_model, max_seq_length)

        self.transformer = nn.Transformer(
            d_model=d_model,
            nhead=nhead,
            num_encoder_layers=num_encoder_layers,
            num_decoder_layers=num_decoder_layers,
            dim_feedforward=dim_feedforward,
            dropout=dropout,
            batch_first=False
        )

        self.fc_out = nn.Linear(d_model, tgt_vocab_size)
        self.dropout = nn.Dropout(dropout)

    def generate_square_subsequent_mask(self, sz):
        """Generate mask for decoder self-attention"""
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask

    def create_padding_mask(self, seq, pad_idx=0):
        """Create padding mask"""
        return (seq == pad_idx)

    def forward(self, src, tgt, src_mask=None, tgt_mask=None,
                src_padding_mask=None, tgt_padding_mask=None):

        # Embeddings and positional encoding
        src = self.dropout(self.pos_encoder(self.src_embedding(src) * math.sqrt(self.d_model)))
        tgt = self.dropout(self.pos_encoder(self.tgt_embedding(tgt) * math.sqrt(self.d_model)))

        # Generate target mask if not provided
        if tgt_mask is None:
            tgt_mask = self.generate_square_subsequent_mask(tgt.size(0)).to(tgt.device)

        # Forward through transformer
        output = self.transformer(src, tgt, src_mask, tgt_mask, None,
                                src_padding_mask, tgt_padding_mask)

        # Final linear layer
        output = self.fc_out(output)

        return output

# ==================== Training Functions ====================

def collate_fn(batch):
    """Custom collate function for DataLoader"""
    gloss_batch, text_batch = zip(*batch)

    # Pad sequences
    gloss_batch = pad_sequence(gloss_batch, batch_first=False, padding_value=0)
    text_batch = pad_sequence(text_batch, batch_first=False, padding_value=0)

    return gloss_batch, text_batch

def train_epoch(model, dataloader, optimizer, criterion, device):
    """Train for one epoch"""
    model.train()
    total_loss = 0

    for batch_idx, (src, tgt) in enumerate(dataloader):
        src = src.to(device)
        tgt = tgt.to(device)

        # Skip batch if too small
        if src.size(1) < 2:
            continue

        # Prepare target input and output
        tgt_input = tgt[:-1, :]
        tgt_output = tgt[1:, :]

        # Create masks - transpose for batch_first=False
        src_seq = src.transpose(0, 1)
        tgt_seq = tgt_input.transpose(0, 1)

        src_padding_mask = model.create_padding_mask(src_seq)
        tgt_padding_mask = model.create_padding_mask(tgt_seq)

        # Forward pass
        optimizer.zero_grad()
        output = model(src, tgt_input,
                      src_padding_mask=src_padding_mask,
                      tgt_padding_mask=tgt_padding_mask)

        # Calculate loss
        loss = criterion(output.reshape(-1, output.shape[-1]),
                        tgt_output.reshape(-1))

        # Backward pass
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()

        total_loss += loss.item()

    return total_loss / max(len(dataloader), 1)

def evaluate(model, dataloader, criterion, device):
    """Evaluate the model"""
    model.eval()
    total_loss = 0

    with torch.no_grad():
        for src, tgt in dataloader:
            src = src.to(device)
            tgt = tgt.to(device)

            # Skip batch if too small
            if src.size(1) < 2:
                continue

            tgt_input = tgt[:-1, :]
            tgt_output = tgt[1:, :]

            # Create masks - transpose for batch_first=False
            src_seq = src.transpose(0, 1)
            tgt_seq = tgt_input.transpose(0, 1)

            src_padding_mask = model.create_padding_mask(src_seq)
            tgt_padding_mask = model.create_padding_mask(tgt_seq)

            output = model(src, tgt_input,
                          src_padding_mask=src_padding_mask,
                          tgt_padding_mask=tgt_padding_mask)

            loss = criterion(output.reshape(-1, output.shape[-1]),
                           tgt_output.reshape(-1))

            total_loss += loss.item()

    return total_loss / max(len(dataloader), 1)

# ==================== Inference ====================

def greedy_decode(model, src, max_len, start_symbol, end_symbol, device):
    """Greedy decoding for inference"""
    model.eval()

    src = src.to(device)
    memory = model.transformer.encoder(
        model.dropout(model.pos_encoder(model.src_embedding(src) * math.sqrt(model.d_model)))
    )

    ys = torch.ones(1, 1).fill_(start_symbol).type(torch.long).to(device)

    for _ in range(max_len - 1):
        tgt = model.dropout(model.pos_encoder(model.tgt_embedding(ys) * math.sqrt(model.d_model)))
        tgt_mask = model.generate_square_subsequent_mask(ys.size(0)).to(device)

        out = model.transformer.decoder(tgt, memory, tgt_mask)
        out = model.fc_out(out)

        prob = out[-1, :, :]
        _, next_word = torch.max(prob, dim=1)
        next_word = next_word.item()

        ys = torch.cat([ys, torch.ones(1, 1).fill_(next_word).type(torch.long).to(device)], dim=0)

        if next_word == end_symbol:
            break

    return ys

def translate_sentence(model, sentence, dataset, device, max_len=50):
    """Translate a single sentence"""
    model.eval()

    # Encode the input sentence
    src = dataset.encode_sentence(sentence, dataset.gloss_to_idx).unsqueeze(1)

    # Get start and end symbols
    start_symbol = dataset.text_to_idx['<sos>']
    end_symbol = dataset.text_to_idx['<eos>']

    # Decode
    tgt_indices = greedy_decode(model, src, max_len, start_symbol, end_symbol, device)

    # Convert indices to words
    tgt_words = []
    for idx in tgt_indices.squeeze():
        idx = idx.item()
        if idx == end_symbol:
            break
        if idx not in [dataset.text_to_idx['<sos>'], dataset.text_to_idx['<pad>']]:
            tgt_words.append(dataset.idx_to_text.get(idx, '<unk>'))

    return ' '.join(tgt_words)

# ==================== Evaluation Metrics ====================

class MetricsCalculator:
    """Calculate various translation metrics"""

    @staticmethod
    def calculate_bleu(references, hypotheses, n_gram=4):
        """Calculate BLEU scores for n-grams 1 to n_gram"""
        smoothing = SmoothingFunction().method1
        bleu_scores = {}

        # Tokenize if needed
        refs_tokenized = [[ref.lower().split()] for ref in references]
        hyps_tokenized = [hyp.lower().split() for hyp in hypotheses]

        # Calculate individual BLEU scores
        for n in range(1, n_gram + 1):
            weights = tuple([1/n] * n + [0] * (4 - n))
            score = corpus_bleu(refs_tokenized, hyps_tokenized,
                              weights=weights, smoothing_function=smoothing)
            bleu_scores[f'BLEU-{n}'] = score * 100

        return bleu_scores

    @staticmethod
    def calculate_meteor(references, hypotheses):
        """Calculate METEOR score"""
        scores = []
        for ref, hyp in zip(references, hypotheses):
            score = meteor_score([ref.lower().split()], hyp.lower().split())
            scores.append(score)
        return np.mean(scores) * 100

    @staticmethod
    def calculate_wer(references, hypotheses):
        """Calculate Word Error Rate"""
        wer_scores = []
        for ref, hyp in zip(references, hypotheses):
            wer = jiwer.wer(ref.lower(), hyp.lower())
            wer_scores.append(wer)
        return np.mean(wer_scores) * 100

    @staticmethod
    def calculate_rouge(references, hypotheses):
        """Calculate ROUGE scores"""
        scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'],
                                         use_stemmer=True)

        rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}

        for ref, hyp in zip(references, hypotheses):
            scores = scorer.score(ref.lower(), hyp.lower())
            for key in rouge_scores.keys():
                rouge_scores[key].append(scores[key].fmeasure)

        # Average scores
        avg_scores = {}
        for key in rouge_scores.keys():
            avg_scores[key.upper()] = np.mean(rouge_scores[key]) * 100

        return avg_scores

# ==================== Main Training Script ====================

def load_signum_data(gloss_file, text_file):
    """Load SIGNUM dataset from files"""
    gloss_sentences = []
    text_sentences = []

    # Read gloss sentences
    with open(gloss_file, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split('\t')
            if len(parts) == 2:
                gloss = parts[1].replace('X-right', '').replace('X-left', '')
                gloss = gloss.replace('X-middle', '').strip()
                gloss = re.sub(r'[+\-]', ' ', gloss)
                gloss_sentences.append(gloss)

    # Read text sentences
    with open(text_file, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split('\t')
            if len(parts) == 2:
                text = parts[1].strip()
                text_sentences.append(text)

    return gloss_sentences, text_sentences

def train_and_evaluate(gloss_sentences, text_sentences,
                       epochs=50, batch_size=32, learning_rate=0.0001,
                       device='cuda' if torch.cuda.is_available() else 'cpu'):
    """Main training and evaluation function"""

    print(f"Using device: {device}")
    print(f"Total sentences: {len(gloss_sentences)}")

    # Split data (80% train, 10% val, 10% test)
    n = len(gloss_sentences)
    indices = list(range(n))
    random.shuffle(indices)

    train_size = int(0.8 * n)
    val_size = int(0.1 * n)

    train_indices = indices[:train_size]
    val_indices = indices[train_size:train_size + val_size]
    test_indices = indices[train_size + val_size:]

    # Create datasets
    train_gloss = [gloss_sentences[i] for i in train_indices]
    train_text = [text_sentences[i] for i in train_indices]

    val_gloss = [gloss_sentences[i] for i in val_indices]
    val_text = [text_sentences[i] for i in val_indices]

    test_gloss = [gloss_sentences[i] for i in test_indices]
    test_text = [text_sentences[i] for i in test_indices]

    # Create dataset objects
    train_dataset = SignumDataset(train_gloss, train_text)
    val_dataset = SignumDataset(val_gloss, val_text,
                                train_dataset.vocab_gloss,
                                train_dataset.vocab_text)
    test_dataset = SignumDataset(test_gloss, test_text,
                                train_dataset.vocab_gloss,
                                train_dataset.vocab_text)

    # Create dataloaders with drop_last=True to avoid batch size issues
    train_loader = DataLoader(train_dataset, batch_size=batch_size,
                            shuffle=True, collate_fn=collate_fn, drop_last=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size,
                          shuffle=False, collate_fn=collate_fn, drop_last=False)
    test_loader = DataLoader(test_dataset, batch_size=1,
                           shuffle=False, collate_fn=collate_fn)

    # Initialize model
    model = TransformerSLT(
        src_vocab_size=len(train_dataset.vocab_gloss),
        tgt_vocab_size=len(train_dataset.vocab_text),
        d_model=256,
        nhead=8,
        num_encoder_layers=3,
        num_decoder_layers=3,
        dim_feedforward=1024,
        dropout=0.1
    ).to(device)

    # Loss and optimizer
    criterion = nn.CrossEntropyLoss(ignore_index=0)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)

    # Training loop
    print("\nStarting training...")
    best_val_loss = float('inf')

    for epoch in range(epochs):
        train_loss = train_epoch(model, train_loader, optimizer, criterion, device)
        val_loss = evaluate(model, val_loader, criterion, device)

        scheduler.step(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), 'best_model.pt')

        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs} - Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

    # Load best model
    model.load_state_dict(torch.load('best_model.pt'))

    # Generate translations for test set
    print("\nGenerating translations for test set...")
    hypotheses = []
    references = []

    for i, (gloss, text) in enumerate(zip(test_gloss, test_text)):
        try:
            translation = translate_sentence(model, gloss, test_dataset, device)
            hypotheses.append(translation)
            references.append(text)

            if i < 5:  # Show first 5 examples
                print(f"\nExample {i+1}:")
                print(f"Gloss: {gloss}")
                print(f"Reference: {text}")
                print(f"Translation: {translation}")
        except Exception as e:
            print(f"Error translating sentence {i}: {e}")
            hypotheses.append("")
            references.append(text)

    # Calculate metrics
    print("\nCalculating evaluation metrics...")
    metrics = MetricsCalculator()

    # Filter out empty translations for metrics
    valid_pairs = [(ref, hyp) for ref, hyp in zip(references, hypotheses) if hyp]
    if valid_pairs:
        valid_references, valid_hypotheses = zip(*valid_pairs)
    else:
        print("No valid translations generated!")
        return model, test_dataset, {}

    # BLEU scores
    bleu_scores = metrics.calculate_bleu(valid_references, valid_hypotheses)

    # METEOR score
    meteor = metrics.calculate_meteor(valid_references, valid_hypotheses)

    # WER
    wer = metrics.calculate_wer(valid_references, valid_hypotheses)

    # ROUGE scores
    rouge_scores = metrics.calculate_rouge(valid_references, valid_hypotheses)

    # Print results
    print("\n" + "="*50)
    print("EVALUATION RESULTS")
    print("="*50)

    for bleu_name, score in bleu_scores.items():
        print(f"{bleu_name}: {score:.2f}")

    print(f"METEOR: {meteor:.2f}")
    print(f"WER: {wer:.2f}")

    for rouge_name, score in rouge_scores.items():
        print(f"{rouge_name}: {score:.2f}")

    return model, test_dataset, {
        **bleu_scores,
        'METEOR': meteor,
        'WER': wer,
        **rouge_scores
    }

# ==================== Example Usage ====================

if __name__ == "__main__":
    # Load your data files
    gloss_file = "signum_sents_anno_eng.txt"
    text_file = "signum_sents_trans_eng.txt"

    # Load data
    gloss_sentences, text_sentences = load_signum_data(gloss_file, text_file)

    # Train and evaluate
    model, dataset, metrics = train_and_evaluate(
        gloss_sentences,
        text_sentences,
        epochs=50,
        batch_size=32,
        learning_rate=0.0001
    )

    print("\nTraining complete!")
    print("\nFinal Metrics Summary:")
    for metric_name, score in metrics.items():
        print(f"{metric_name}: {score:.2f}")
```

```python
#signum text to gloss t5
import pandas as pd
from datasets import Dataset
import torch
from transformers import (
    T5ForConditionalGeneration,
    T5Tokenizer,
    Seq2SeqTrainingArguments,
    Seq2SeqTrainer,
    EarlyStoppingCallback,
    DataCollatorForSeq2Seq
)
from sklearn.metrics import accuracy_score
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction
from sklearn.model_selection import train_test_split
import nltk
from nltk.tokenize import word_tokenize
import random
import numpy as np
import logging
import jiwer
from rouge_score import rouge_scorer
import re

# ─── Setup ─────────────────────────────────────────────────────────────────────
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    handlers=[logging.StreamHandler(), logging.FileHandler("training_text_to_gloss.log")]
)
logger = logging.getLogger(__name__)

nltk.download("punkt")
nltk.download("wordnet")

# ─── Sign Language Gloss Specific Helpers ─────────────────────────────────────

def parse_signum_file(filepath, file_type='gloss'):
    """
    Parse Signum dataset files flexibly.
    Handles various delimiters and formats.
    """
    data = []

    with open(filepath, 'r', encoding='utf-8') as f:
        for line_num, line in enumerate(f, 1):
            line = line.strip()
            if not line or line.startswith('#'):  # Skip empty lines and comments
                continue

            # Try tab delimiter first (most common)
            if '\t' in line:
                parts = line.split('\t', 1)  # Split only on first tab
                if len(parts) == 2:
                    data.append({
                        'id': parts[0].strip(),
                        file_type: parts[1].strip()
                    })
                    continue

            # Try other delimiters
            for delimiter in ['|', '::']:
                if delimiter in line:
                    parts = line.split(delimiter, 1)
                    if len(parts) == 2:
                        data.append({
                            'id': parts[0].strip(),
                            file_type: parts[1].strip()
                        })
                        break
            else:
                # Try to extract ID from the beginning (space-separated)
                match = re.match(r'^(\S+)\s+(.+)$', line)
                if match:
                    data.append({
                        'id': match.group(1),
                        file_type: match.group(2)
                    })
                else:
                    logger.warning(f"Could not parse line {line_num}: {line[:50]}...")

    return data

def preserve_gloss_notation(text):
    """
    Preserve important sign language gloss notation while doing minimal cleaning.
    English glosses often use CAPITALS for signs and special notation.
    """
    if not text:
        return ""

    # Preserve the original text structure
    text = text.strip()

    # Normalize only excessive whitespace (keep single spaces)
    text = re.sub(r'\s+', ' ', text)

    # DO NOT lowercase - capitals indicate signs in glosses
    # DO NOT remove underscores - they connect multi-word signs
    # DO NOT remove hyphens - they may connect compound signs
    # DO NOT remove special characters like +, /, #, : - they have linguistic meaning

    # Only remove truly problematic characters
    text = text.replace('\u200b', '')  # Remove zero-width spaces
    text = text.replace('\ufeff', '')  # Remove BOM
    text = text.replace('\xa0', ' ')   # Replace non-breaking spaces with regular spaces

    return text

def process_english_text(text):
    """
    Process natural English text for input.
    Unlike glosses, we can normalize this more.
    """
    if not text:
        return ""

    text = text.strip()
    text = re.sub(r'\s+', ' ', text)

    # For natural text, we can lowercase and normalize punctuation
    # But keep it readable
    text = text.lower()

    # Remove excessive punctuation but keep sentence boundaries
    text = re.sub(r'[.!?]+', '.', text)
    text = re.sub(r',+', ',', text)

    return text

def process_english_gloss(gloss):
    """
    Process English sign language glosses as output.
    Preserves linguistic markers while preparing for generation.
    """
    # Preserve gloss notation
    gloss = preserve_gloss_notation(gloss)

    # Common English gloss patterns to preserve:
    # - CAPITALS for signs (e.g., BOOK, GIVE, YESTERDAY)
    # - Hyphens for compound signs (e.g., LOOK-AT, GIVE-UP)
    # - Underscores for multi-word concepts (e.g., SIGN_LANGUAGE)
    # - IX/INDEX for pointing/indexing
    # - CL: for classifiers
    # - ++ for repetition (e.g., WALK++)
    # - Numbers for incorporation (e.g., THREE-WEEKS)
    # - # for fingerspelling (e.g., #JOHN)
    # - Parentheses for non-manual markers (e.g., (head-shake))

    return gloss

def augment_text_gloss_pair(text, gloss, p=0.15):
    """
    Careful augmentation for text-to-gloss pairs.
    Creates variations while maintaining meaning.
    """
    if random.random() > p:
        return text, gloss

    augmentation_type = random.choice(['paraphrase_text', 'simplify_gloss', 'none'])

    if augmentation_type == 'none':
        return text, gloss

    if augmentation_type == 'paraphrase_text':
        # Simple text paraphrases that don't change meaning
        text_replacements = {
            "i went": "i traveled",
            "i bought": "i purchased",
            "the store": "the shop",
            "yesterday": "the day before",
            "tomorrow": "the next day",
            "want to": "would like to",
            "need to": "have to"
        }

        text_lower = text.lower()
        for old, new in text_replacements.items():
            if old in text_lower:
                text = text_lower.replace(old, new)
                return text, gloss

    elif augmentation_type == 'simplify_gloss':
        # Remove redundant elements in gloss that are optional
        gloss_tokens = gloss.split()
        # Articles are often omitted in glosses
        optional_elements = ['THE', 'A', 'AN']
        filtered = [t for t in gloss_tokens if t not in optional_elements]
        if len(filtered) < len(gloss_tokens) and len(filtered) > 2:
            return text, ' '.join(filtered)

    return text, gloss

# ─── Evaluation Metrics ────────────────────────────────────────────────────────

def calculate_gloss_accuracy(reference_gloss, predicted_gloss):
    """
    Calculate accuracy specific to gloss generation.
    Considers sign-level accuracy and structure.
    """
    ref_signs = [w for w in reference_gloss.split() if w.isupper()]
    pred_signs = [w for w in predicted_gloss.split() if w.isupper()]

    if not ref_signs:
        return 1.0 if not pred_signs else 0.0

    # Calculate sign-level accuracy
    correct = 0
    for i, sign in enumerate(ref_signs):
        if i < len(pred_signs) and pred_signs[i] == sign:
            correct += 1

    position_accuracy = correct / max(len(ref_signs), len(pred_signs))

    # Calculate sign coverage (did we generate all necessary signs?)
    ref_set = set(ref_signs)
    pred_set = set(pred_signs)
    coverage = len(ref_set.intersection(pred_set)) / len(ref_set) if ref_set else 0

    # Combined score
    return (position_accuracy + coverage) / 2

def calculate_bleu_scores(reference, hypothesis):
    """Calculate BLEU scores for gloss generation."""
    # For glosses, we want exact matches more than for natural language
    r = reference.split()  # Don't lowercase for glosses
    h = hypothesis.split()

    # Use smoothing for short sentences
    smoothie = SmoothingFunction().method1

    weights_list = [
        (1, 0, 0, 0),      # BLEU-1
        (0.5, 0.5, 0, 0),  # BLEU-2
        (1/3, 1/3, 1/3, 0), # BLEU-3
        (0.25, 0.25, 0.25, 0.25), # BLEU-4
    ]

    scores = {}
    for i, w in enumerate(weights_list, start=1):
        try:
            score = corpus_bleu([[r]], [h], weights=w, smoothing_function=smoothie)
            scores[f"bleu_{i}"] = score
        except:
            scores[f"bleu_{i}"] = 0.0

    return scores

def calculate_wer(reference, hypothesis):
    """Calculate Word Error Rate for glosses."""
    try:
        # For glosses, case matters
        return jiwer.wer(reference, hypothesis)
    except:
        return 1.0

def calculate_rouge_scores(reference, hypothesis):
    """Calculate ROUGE scores for gloss generation."""
    try:
        scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=False)
        sc = scorer.score(reference, hypothesis)
        return {
            'rouge1_fmeasure': sc['rouge1'].fmeasure,
            'rouge2_fmeasure': sc['rouge2'].fmeasure,
            'rougeL_fmeasure': sc['rougeL'].fmeasure
        }
    except:
        return {
            'rouge1_fmeasure': 0.0,
            'rouge2_fmeasure': 0.0,
            'rougeL_fmeasure': 0.0
        }

def set_seed(seed=42):
    """Set random seeds for reproducibility."""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    logger.info(f"Seed set to {seed}")

# ─── Main ──────────────────────────────────────────────────────────────────────

if __name__ == "__main__":
    set_seed()

    # Parse Signum dataset files
    anno_path = 'signum_sents_anno_eng.txt'  # Glosses
    trans_path = 'signum_sents_trans_eng.txt'  # Natural English text

    logger.info("Parsing Signum dataset files for TEXT-TO-GLOSS translation...")

    # Load gloss data (this will be our TARGET)
    try:
        gloss_data = parse_signum_file(anno_path, 'gloss')
        logger.info(f"Loaded {len(gloss_data)} gloss entries (targets)")
    except Exception as e:
        logger.error(f"Error loading gloss data: {e}")
        raise

    # Load translation data (this will be our INPUT)
    try:
        trans_data = parse_signum_file(trans_path, 'translation')
        logger.info(f"Loaded {len(trans_data)} translation entries (inputs)")
    except Exception as e:
        logger.error(f"Error loading translation data: {e}")
        raise

    # Merge datasets
    gloss_df = pd.DataFrame(gloss_data)
    trans_df = pd.DataFrame(trans_data)
    df = pd.merge(gloss_df, trans_df, on='id', how='inner')

    # REVERSED: translation is input, gloss is output
    df['input_text'] = df['translation'].apply(process_english_text)
    df['target_gloss'] = df['gloss'].apply(process_english_gloss)

    # Remove empty entries
    df = df[df['input_text'].str.len() > 0]
    df = df[df['target_gloss'].str.len() > 0]

    logger.info(f"{len(df)} valid parallel examples after merging")

    # Log some examples to verify data
    logger.info("\nSample data (TEXT -> GLOSS):")
    for i in range(min(3, len(df))):
        logger.info(f"  Input text: {df.iloc[i]['input_text'][:100]}")
        logger.info(f"  Target gloss: {df.iloc[i]['target_gloss'][:100]}")
        logger.info("  ---")

    # Careful augmentation
    augmented_rows = []
    for _, row in df.iterrows():
        augmented_rows.append({
            'id': row['id'],
            'input_text': row['input_text'],
            'target_gloss': row['target_gloss']
        })

        # Augment only 15% of data
        aug_text, aug_gloss = augment_text_gloss_pair(
            row['input_text'],
            row['target_gloss'],
            p=0.15
        )

        if aug_text != row['input_text'] or aug_gloss != row['target_gloss']:
            augmented_rows.append({
                'id': row['id'] + '_aug',
                'input_text': aug_text,
                'target_gloss': aug_gloss
            })

    df = pd.DataFrame(augmented_rows)
    logger.info(f"After augmentation: {len(df)} examples")

    # Split dataset
    train_df, tmp_df = train_test_split(df, test_size=0.2, random_state=42)
    eval_df, test_df = train_test_split(tmp_df, test_size=0.5, random_state=42)

    logger.info(f"Splits — train: {len(train_df)}, eval: {len(eval_df)}, test: {len(test_df)}")

    # Convert to HuggingFace datasets
    train_ds = Dataset.from_pandas(train_df)
    eval_ds = Dataset.from_pandas(eval_df)
    test_ds = Dataset.from_pandas(test_df)

    # Use T5-base for text-to-gloss translation
    model_name = "t5-base"

    logger.info(f"Loading model: {model_name}")
    tokenizer = T5Tokenizer.from_pretrained(model_name, use_fast=False, legacy=False)
    model = T5ForConditionalGeneration.from_pretrained(model_name)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    logger.info(f"Using device: {device}")
    model.to(device)

    # Preprocessing function for TEXT to GLOSS
    def preprocess_function(examples):
        # Task prefix for text-to-gloss
        inputs = [f"convert to sign language gloss: {text}" for text in examples["input_text"]]

        # Tokenize inputs (natural English text)
        model_inputs = tokenizer(
            inputs,
            max_length=128,
            truncation=True,
            padding="max_length",
            return_tensors="pt"
        )

        # Tokenize targets (glosses)
        with tokenizer.as_target_tokenizer():
            labels = tokenizer(
                examples["target_gloss"],
                max_length=128,
                truncation=True,
                padding="max_length",
                return_tensors="pt"
            )

        # Replace padding token id with -100 for loss calculation
        labels["input_ids"][labels["input_ids"] == tokenizer.pad_token_id] = -100

        model_inputs["labels"] = labels["input_ids"]

        # Convert tensors back to lists for dataset storage
        model_inputs = {k: v.tolist() for k, v in model_inputs.items()}

        return model_inputs

    # Apply preprocessing
    logger.info("Preprocessing datasets for text-to-gloss...")
    train_ds = train_ds.map(
        preprocess_function,
        batched=True,
        remove_columns=["id", "input_text", "target_gloss"],
        desc="Preprocessing train dataset"
    )
    eval_ds = eval_ds.map(
        preprocess_function,
        batched=True,
        remove_columns=["id", "input_text", "target_gloss"],
        desc="Preprocessing eval dataset"
    )
    test_ds = test_ds.map(
        preprocess_function,
        batched=True,
        remove_columns=["id", "input_text", "target_gloss"],
        desc="Preprocessing test dataset"
    )

    # Set format for PyTorch
    train_ds.set_format("torch")
    eval_ds.set_format("torch")
    test_ds.set_format("torch")

    # Data collator
    data_collator = DataCollatorForSeq2Seq(
        tokenizer,
        model=model,
        padding=True,
        max_length=128
    )

    # Training arguments optimized for text-to-gloss generation
    batch_size = 4 if torch.cuda.is_available() else 2

    training_args = Seq2SeqTrainingArguments(
        output_dir="./results_t5_text_to_gloss",
        eval_strategy="steps",
        eval_steps=50,
        save_steps=100,
        save_total_limit=3,
        num_train_epochs=8,  # More epochs for harder task
        learning_rate=5e-4,  # Slightly higher for text-to-gloss
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        gradient_accumulation_steps=4,
        warmup_ratio=0.1,
        weight_decay=0.01,
        logging_dir="./logs_t5_text_to_gloss",
        logging_steps=10,
        logging_first_step=True,
        report_to="none",
        predict_with_generate=True,
        generation_max_length=128,
        generation_num_beams=5,  # More beams for gloss generation
        load_best_model_at_end=True,
        metric_for_best_model="gloss_accuracy",
        greater_is_better=True,
        fp16=torch.cuda.is_available(),
        seed=42,
        dataloader_drop_last=True,
    )

    # Compute metrics function for gloss generation
    def compute_metrics(eval_pred):
        predictions, labels = eval_pred

        # Handle both logits and token ids
        if isinstance(predictions, tuple):
            predictions = predictions[0]

        if predictions.ndim == 3:
            predictions = np.argmax(predictions, axis=-1)

        # Filter out invalid token IDs before decoding
        vocab_size = tokenizer.vocab_size

        # Clean predictions
        predictions = np.where(
            (predictions >= 0) & (predictions < vocab_size),
            predictions,
            tokenizer.pad_token_id
        )

        # Replace -100 in labels with pad token for decoding
        labels = np.where(labels != -100, labels, tokenizer.pad_token_id)

        # Clean labels
        labels = np.where(
            (labels >= 0) & (labels < vocab_size),
            labels,
            tokenizer.pad_token_id
        )

        # Decode predictions and labels
        decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
        decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

        # Clean up decoded text
        decoded_preds = [pred.strip() for pred in decoded_preds]
        decoded_labels = [label.strip() for label in decoded_labels]

        # Calculate metrics
        metrics = {
            'gloss_accuracy': [],
            'bleu_1': [],
            'bleu_2': [],
            'bleu_3': [],
            'bleu_4': [],
            'wer': [],
            'rouge1_fmeasure': [],
            'rouge2_fmeasure': [],
            'rougeL_fmeasure': [],
        }

        for ref, hyp in zip(decoded_labels, decoded_preds):
            if ref and hyp:  # Only compute if both are non-empty
                # Gloss-specific accuracy
                metrics['gloss_accuracy'].append(calculate_gloss_accuracy(ref, hyp))

                # BLEU scores
                bleu = calculate_bleu_scores(ref, hyp)
                for k, v in bleu.items():
                    metrics[k].append(v)

                # WER
                metrics['wer'].append(calculate_wer(ref, hyp))

                # ROUGE scores
                rouge = calculate_rouge_scores(ref, hyp)
                for k, v in rouge.items():
                    metrics[k].append(v)

        # Return averaged metrics
        return {k: float(np.mean(v)) if v else 0.0 for k, v in metrics.items()}

    # Initialize trainer
    trainer = Seq2SeqTrainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=eval_ds,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]
    )

    # Train the model
    logger.info("Starting training for TEXT-TO-GLOSS translation...")
    try:
        train_result = trainer.train()
        logger.info(f"Training completed. Final loss: {train_result.training_loss}")

        # Save training metrics
        trainer.log_metrics("train", train_result.metrics)
        trainer.save_metrics("train", train_result.metrics)

    except Exception as e:
        logger.error(f"Error during training: {e}")
        raise

    # Evaluate on test set
    logger.info("Evaluating on test set...")
    test_metrics = trainer.evaluate(test_ds, metric_key_prefix="test")
    logger.info(f"Test metrics: {test_metrics}")

    # Save test metrics
    trainer.log_metrics("test", test_metrics)
    trainer.save_metrics("test", test_metrics)

    # Save the model and tokenizer
    output_dir = "./fine_tuned_t5_text_to_gloss"
    model.save_pretrained(output_dir)
    tokenizer.save_pretrained(output_dir)
    logger.info(f"Model and tokenizer saved to {output_dir}")

    # Test generation function
    def text_to_gloss(text):
        """Convert natural English text to sign language gloss."""
        inputs = tokenizer(
            f"convert to sign language gloss: {text.lower()}",
            return_tensors="pt",
            max_length=128,
            truncation=True
        ).to(device)

        outputs = model.generate(
            **inputs,
            max_length=128,
            num_beams=5,
            early_stopping=True,
            no_repeat_ngram_size=2,  # Avoid repetition in glosses
            temperature=0.9,  # Slight randomness for natural glosses
        )

        gloss = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return gloss

    # Example generations with detailed evaluation
    logger.info("\n=== Example TEXT-TO-GLOSS Translations ===")
    test_samples = test_df.sample(min(10, len(test_df)))

    print("\n" + "="*80)
    print("TEXT-TO-GLOSS GENERATION EXAMPLES")
    print("="*80)

    for idx, (_, row) in enumerate(test_samples.iterrows(), 1):
        text = row['input_text']
        reference_gloss = row['target_gloss']
        predicted_gloss = text_to_gloss(text)

        # Calculate metrics for this example
        gloss_acc = calculate_gloss_accuracy(reference_gloss, predicted_gloss)
        bleu = calculate_bleu_scores(reference_gloss, predicted_gloss)
        wer = calculate_wer(reference_gloss, predicted_gloss)

        print(f"\nExample {idx}:")
        print(f"Input Text: {text}")
        print(f"Reference Gloss: {reference_gloss}")
        print(f"Predicted Gloss: {predicted_gloss}")
        print(f"Metrics: Gloss Acc: {gloss_acc:.3f} | BLEU-4: {bleu['bleu_4']:.3f} | WER: {wer:.3f}")
        print("-" * 80)

    # Print final summary
    print("\n" + "="*80)
    print("TRAINING COMPLETE - TEXT TO GLOSS MODEL")
    print("="*80)
    print(f"Model saved to: {output_dir}")
    print(f"\nTest metrics:")
    for key, value in test_metrics.items():
        if 'test_' in key:
            metric_name = key.replace('test_', '')
            print(f"  {metric_name}: {value:.4f}")
    print("="*80)
    print("\nModel is ready to convert English text to sign language glosses!")
    print("Usage: gloss = text_to_gloss('your english text here')")
    print("="*80)
```
